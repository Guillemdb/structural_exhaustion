## 1. Continuum Problem Signature

### Definition 1.1 (Legal Continuum Signature)
A continuum problem signature is
$$
\mathfrak P=(X,\mathcal F_X,\mu,\mathcal I,q,E_q,\mathcal L_W,\mathcal B,\Sigma),
$$
where:

- $(X,\mathcal F_X,\mu)$ is a presented state space with reference measure fixed before solving the dynamics.
- $\mathcal I=\{O_a:X\to V_a\}_{a\in A}$ is a public measurable observable interface and
  $$
  \mathcal F_{\rm obs}=\sigma\{O_a^{-1}(B):a\in A,\ B\in\mathcal B_{V_a}\}.
  $$
- $q:X\to Q$ is a represented quotient, and $E_q=L_XU-UL_Q$, $Uf=f\circ q$, is its lumpability defect.
- $\mathcal L_W\ge0$ is a represented defect geometry: a nonnegative self-adjoint operator or closed symmetric form used to route $E_q$.
- $\mathcal B=(\mathcal R,{\rm cost},b)$ is a budget satisfying B1-B4 from Section 2.
- $\Sigma\in\mathcal F_{\rm obs}$, or else its saturated hull $q^{-1}q(\Sigma)$, is the target or forbidden set.

The signature is legal when:

1. $\mu$ and $\mathcal I$ are presentation-fixed, not trajectory laws or solution-dependent gauges.
2. $q$ and $\Sigma$ are public objects, represented in the observable algebra up to null sets.
3. $L$ is the generator of a quasi-regular coercive form satisfying the sector condition stated in Section 3.
4. The structural-gradient, quotient, and lift subspaces used by the decomposition are closed, or else the notebook's boundary-defect repair theorem is invoked to close them on a stratified space.
5. All costs used to declare affordability are measured by $\mathcal B$, not by an implicit external regularity assumption.
6. Conditional expectations, gauge projections, noncommutative observable reductions, and expectation-gauge fluctuation transcripts are represented objects with their construction or approximation cost charged to $\mathcal B$ whenever those rows are invoked.

### Theorem 1.2 (Registration Theorem)
A legal continuum signature places the problem in the framework's continuum interface. The base consequences are the existence of the represented generator/form, the represented defect $E_q$, the represented defect geometry $\mathcal L_W$, and the budget capacity of $\Sigma$. Each later metatheorem triggers only after its own additional interface hypotheses are supplied: closed range for exhaustiveness, finite resistance for routing, zero capacity for polarity, scale-stability for reaching, SCRC data for boundary-defect closure, conservative carrier data for the fast-track, elliptic tail and gauge-cokernel data for constraint-tail closure, flux numerical-radius data for sign certification, or Wiener/law-space data for stochastic extensions, and conditional-expectation/fluctuation-resistance data for noncommutative gauge or expectation-gauge rows.

**Proof.** Legality supplies the common objects used by all later modules: a closed generator form, public represented subspaces, a defect, a geometry, a target, and a budget. Each later theorem is a conditional implication whose hypotheses name an additional certificate on this registered interface: a gap, quotient, resistance, capacity, ledger, carrier, or realization transcript. The proof of each module reads only the registered objects and its named certificate, so no module may use an unregistered solver or a conclusion from a later row. Without the local certificate, the row remains at registration and no stronger module conclusion is triggered. $\square$


## 2. Budgets, Affordability, and Capacity

### Definition 2.1 (Budget Structure)
A budget structure is <span class="math">$\mathcal B=(\mathcal R,{\rm cost},b,\mathfrak C_{\mathcal B})$</span>, where the resource space is an ordered commutative monoid
$$
\mathcal R=(C,\preceq,\oplus,\mathbf 0),
$$
where $\preceq$ is the “no more resource than” order, $\oplus$ accumulates costs, and $\mathfrak C_{\mathcal B}$ is the budget transfer class of allowed overheads. The map ${\rm cost}:\mathsf{Constr}(I_n)\to C$ charges represented observables, quotients, potentials, finite approximants, and certificates, and $b(n)$ is the size-indexed ceiling. The later flux sections use $\mathfrak C(\tau)$ for a dissipation density; that analytic density is not the budget transfer class $\mathfrak C_{\mathcal B}$.

A constructor $\gamma$ is affordable when
$$
{\rm Constr}_{\mathcal B}(\gamma)=1
\quad\Longleftrightarrow\quad
\exists\gamma'\sim\gamma\text{ with }{\rm cost}(\gamma')\preceq b(n).
$$
The affordable observable algebra is
$$
\mathcal F^{\mathcal B}_{I_n}=\sigma\{\gamma:{\rm cost}(\gamma)\preceq b(n)\}\subseteq\mathcal F_{I_n}.
$$

### Axioms B1-B4
The budget axioms are:

- **B1:** $\preceq$ is a decidable monotone ceiling order and $\oplus$ is monotone.
- **B2:** bounded composition of affordable constructors remains in the same budget transfer class after a class-bounded ceiling enlargement.
- **B3:** affordability depends on the represented object, not on syntax.
- **B4:** static affordability and dynamic stopping levels are comparable in the same resource order.

### Definition 2.2 (Budget Capacity)
For a quasi-regular form $(\mathcal E,D(\mathcal E))$ and target $A\subset X$, define the budget capacity
$$
{\rm Cap}_{\mathcal B}(A)
=\inf\{\mathcal E(f,f): f\in D(\mathcal E),\ f\ge 1\text{ on a neighborhood of }A,\ f\text{ affordable at }\mathcal B\}.
$$
If the problem uses a family of budgets, for example a moment-indexed family, write ${\rm Cap}^{(p)}_{\mathcal B}$ for the capacity under the $p$th cost charge.

### Theorem 2.3 (Budget Transfer Closure)
Any structural theorem whose proof uses only B1-B4 transfers unchanged from one budget to any other budget in the same budget transfer class. In particular, affordable observable closure, quotient/lift composition closure, projection ceilings restricted to $\mathcal F^{\mathcal B}$, and capacity comparisons remain valid after replacing the concrete ceiling by any ceiling satisfying B1-B4.

**Proof.** The only cost-sensitive step in the structural proofs is bounded composition. If $\gamma$ is built from affordable subconstructors $\gamma_i$, sub-accumulation gives
$$
{\rm cost}(\gamma)\preceq \bigoplus_i {\rm cost}(\gamma_i)\oplus {\rm overhead}.
$$
B1 compares each input with $b(n)$, B2 absorbs the bounded accumulated overhead into a class-equivalent ceiling, B3 makes the conclusion representation-independent, and B4 keeps the static ceiling compatible with the run envelope. The measure-theoretic identities used by lift and quotient arguments contain no budget-specific fact. $\square$


## 3. Continuum Admissibility and Generator Form

### Standing Hypothesis 3.1 (Continuum Admissibility)
The continuum generator $L$ is represented by a quasi-regular coercive closed form
$$
\mathcal E(u,v)=\mathcal E^{\mathsf s}(u,v)+\mathcal E^{\mathsf a}(u,v)+\mathcal E^{\mathsf b}(u,v)
$$
on $L^2(X,\mu)$, with symmetric part $\mathcal E^{\mathsf s}$, antisymmetric part $\mathcal E^{\mathsf a}$, and nonnegative killing/boundary part $\mathcal E^{\mathsf b}$. The symmetric part is presented in a Beurling-Deny-LeJan representation on the chosen regular or quasi-regular state presentation. The sector condition holds:
$$
|\mathcal E^{\mathsf a}(u,v)|
\le K\,\mathcal E^{\mathsf s}_1(u,u)^{1/2}\mathcal E^{\mathsf s}_1(v,v)^{1/2},
\qquad \mathcal E^{\mathsf s}_1=\mathcal E^{\mathsf s}+\|\cdot\|_2^2.
$$
The associated right process and capacity/polar theory exist. The structural subspaces used below are Hilbert subspaces of the relevant form or edge spaces.

### Theorem 3.2 (Continuum Generator Form)
Under Standing Hypothesis 3.1, the symmetric Markovian part of the admissible continuum generator decomposes into the represented strongly local diffusion part, symmetric jump part, and killing part. Within this Beurling-Deny-LeJan presentation there is no additional symmetric Markovian primitive.

**Proof.** This is the Beurling-Deny-LeJan representation theorem for the declared regular or quasi-regular Dirichlet-form presentation. The theorem gives the three terms and uniqueness of the representing measures/forms on the presentation. Hence any represented symmetric Markovian contribution lies in one of those three represented terms. $\square$

### Theorem 3.3 (Sector Split)
Under the sector condition, $L=L^{\mathsf s}+L^{\mathsf a}+L^{\mathsf b}$ with $L^{\mathsf s}$ self-adjoint on the symmetric form domain, $L^{\mathsf a}$ form-bounded and antisymmetric, and $L^{\mathsf b}$ nonnegative killing or boundary readout.

**Proof.** The sector bound makes $\mathcal E^{\mathsf a}$ a bounded perturbation of $\mathcal E^{\mathsf s}_1$ in the form norm. The Kato-Lions sectorial form theorem gives a closed sectorial form with the same domain, and polarization separates the symmetric, antisymmetric, and killing/boundary parts. $\square$

### Theorem 3.4 (Closed Projection Chain)
If the represented quotient, lift, and structural-gradient subspaces are closed, orthogonal projection onto them is unique and the corresponding channel components are well-defined. If any such range is not closed, projection onto its closure is the available Hilbert-space projection; the difference between the closure and the range is a boundary defect.

**Proof.** Orthogonal projection exists exactly onto closed subspaces of a Hilbert space. Applying the projection theorem to each closed represented subspace gives the decomposition chain. If a subspace is not closed, the projection theorem applies only to its closure, and elements in $\overline S\setminus S$ have no represented channel witness. $\square$


## 4. Continuum Channel Decomposition and Boundary Defect

Let $\mathcal H_{\rm edge}$ be the Hilbert space of antisymmetric currents or edge fields, and let
$$
G:\mathcal P_{\rm str}\to\mathcal H_{\rm edge}
$$
be the structural-gradient map from admissible structural potentials into currents. Write $\mathcal C_{\rm str}={\rm im}\,G$.

### Definition 4.1 (Boundary Defect)
The boundary defect is
$$
\partial_{\rm bdry}\mathcal C_{\rm str}
=\overline{{\rm im}\,G}\setminus {\rm im}\,G.
$$
An element of this set is an $L^2$ or form-limit of represented structural gradients but is not itself the gradient of an admissible structural potential.

### Theorem 4.2 (Directed Exhaustiveness iff Closed Range)
Assume the generator-form and quotient/lift hypotheses of Section 3. The directed part of the continuum channel decomposition has no third category if and only if $G$ has closed range. Equivalently,
$$
\mathcal H_{\rm edge}={\rm im}\,G\oplus({\rm im}\,G)^\perp
$$
with a represented structural-gradient component and a represented residual component if and only if $\overline{{\rm im}\,G}={\rm im}\,G$.

**Proof.** If ${\rm im}\,G$ is closed, the projection theorem gives an orthogonal split onto ${\rm im}\,G$ and its complement. The first part is a represented Caus current; the second is orthogonal to every represented structural gradient and is residual. Conversely, if ${\rm im}\,G$ is not closed, choose $F\in\overline{{\rm im}\,G}\setminus {\rm im}\,G$. Then $F$ is not represented as a structural gradient, but it is not residual either, because it is a limit of structural gradients rather than orthogonal to them. Thus a third directed category exists. $\square$

### Theorem 4.3 (Closed-Range Criterion)
Assume $G$ is closed and densely defined between Hilbert spaces. Then ${\rm im}\,G$ is closed if and only if the structural Laplacian
$$
\Delta_{\rm str}=G^*G
$$
has a spectral gap away from $0$ on $(\ker G)^\perp$. Equivalently, there is $\gamma_{\rm str}>0$ such that
$$
\|\Phi\|^2\le \gamma_{\rm str}^{-1}\|G\Phi\|^2,
\qquad \Phi\in(\ker G)^\perp\cap D(G).
$$

**Proof.** The Hilbert-space closed-range theorem states that ${\rm im}\,G$ is closed iff $G$ is bounded below on $(\ker G)^\perp$ in the graph norm. This lower bound is precisely the Poincare inequality displayed above. Since $\|G\Phi\|^2=\langle G^*G\Phi,\Phi\rangle$, the best such lower bound is the bottom of the nonzero spectrum of $G^*G$. $\square$

### Corollary 4.4 (Regularized Certificate)
For every $\lambda>0$, the regularized normal operator $\Delta_{\rm str}+\lambda I$ is coercive and has a bounded inverse on the Hilbert space. Therefore the ridged certificate
$$
\langle E,(\Delta_{\rm str}+\lambda I)^{-1}E\rangle
$$
is always computable for represented $E$ in the working Hilbert space, with a declared $\lambda$-bias. This is a regularized certificate, not a proof that the unregularized range of $G$ is closed. The unregularized decomposition still requires Theorem 4.3 or the boundary-defect fast-track below.

**Proof.** $\langle(\Delta_{\rm str}+\lambda I)\Phi,\Phi\rangle\ge\lambda\|\Phi\|^2$, so the inverse exists with norm at most $\lambda^{-1}$. This gives a stable ridged solve and a finite certificate. Closedness of ${\rm im}\,G$ is a property of $G$, not of $G^*G+\lambda I$; hence the ridge supplies an approximation with a charged bias but does not discharge the exact closed-range condition. $\square$

The non-closed case is therefore handled as a target-relative quotient, not as an ambient concentration-compactness problem. Section 13 proves the affordable completion theorem used below: any compactified class escaping every affordable window has zero flux through every affordable target shell and is placed in $J_{\rm res}^{0\Sigma}$; only in-window target-visible defect classes remain in the quotient $\mathfrak H_{\rm bdry}^{\Sigma}$. Thus target-completeness is a theorem of the affordable algebra once the target compactification is supplied.

### Definition 4.5 (Target-Complete Boundary-Defect Ledger)
Fix the target $\Sigma$. Let $\widehat{\mathcal H}_{\rm bdry}^{\mathcal B,\Sigma}$ denote the compactified target boundary-channel space generated by the affordable structural closure $\overline{{\rm im}\,G}^{\mathcal B}$, its compactified boundary-defect limits, and the residual zero-target classes produced by the affordable target compactification. A boundary-defect closed ledger is a closed represented subspace or closed convex cone
$$
\mathcal K_{\rm bdry,closed}^{\Sigma}\subseteq
\widehat{\mathcal H}_{\rm bdry}^{\mathcal B,\Sigma}
$$
that contains every certified non-reaching boundary-defect mechanism:

1. represented structural gradients ${\rm im}\,G$;
2. residual zero-target-flux classes $J_{\rm res}^{0\Sigma}$;
3. gauge, symmetry, lift, and pure boundary-readout directions with zero target flux;
4. dyadic-decay classes, rebased carriers, and subthreshold classes;
5. rigid classes reclassified into a closed structural-gradient range or excluded by a Liouville/rigidity certificate;
6. classes killed by represented Pohozaev, Noether, monotonicity, virial, or frequency identities;
7. every represented or compactified boundary-defect class whose relative target capacity is zero.

It is target-complete when the converse holds inside the declared compactified interface: any represented or compactified boundary-defect class with zero target capacity or zero certified normal target flux lies in $\mathcal K_{\rm bdry,closed}^{\Sigma}$. The boundary-defect cohomology is
$$
\mathfrak H_{\rm bdry}^{\Sigma}
=
\widehat{\mathcal H}_{\rm bdry}^{\mathcal B,\Sigma}/\mathcal K_{\rm bdry,closed}^{\Sigma}.
$$
Escaping representatives of compactified limits are first placed in $J_{\rm res}^{0\Sigma}$ by Lemma 13.6; the remaining nonzero quotient classes are therefore in-window and target-visible.

### Theorem 4.6 (Boundary-Defect Fast-Track)
Assume the legal interface of Sections 1-3 and a target-complete boundary-defect ledger. Then exactly one alternative holds.

**Closed alternative.** If
$$
\mathfrak H_{\rm bdry}^{\Sigma}=0,
$$
then every boundary defect lies in one of the ledger generators of Definition 4.5: represented structural gradients, residual zero-flux classes, gauge/symmetry/lift readouts, dyadic/rebased/subthreshold classes, Pohozaev/Noether-killed classes, rigidly excluded classes, or zero-capacity classes. Continuum exhaustiveness is restored relative to $\Sigma$.

**Open alternative.** If
$$
\mathfrak H_{\rm bdry}^{\Sigma}\ne0,
$$
then the compactified interface contains a represented target-visible boundary-defect class with positive target capacity and nonzero certified target flux. With a realization/nonpolarity certificate for that positive-capacity target-visible class, $\Sigma$ is reached.

**Proof.** The ledger is closed, so the quotient is represented. If $\mathfrak H_{\rm bdry}^{\Sigma}=0$, every element of $\widehat{\mathcal H}_{\rm bdry}^{\mathcal B,\Sigma}$ lies in $\mathcal K_{\rm bdry,closed}^{\Sigma}$. Definition 4.5 lists the allowed generators of this ledger. Structural gradients are represented Caus channels by Theorem 4.2; residual zero-flux and escaping classes are placed in $J_{\rm res}^{0\Sigma}$ by Lemma 13.6; zero-capacity classes are non-reaching by Theorem 7.2; the remaining generators enter only through their named gauge, dyadic, rebasing, subthreshold, rigidity, Pohozaev, Noether, virial, or frequency certificates. Closure preserves zero-reach conclusions by lower semicontinuity of capacity and continuity of the represented target-flux functional. Therefore every boundary-defect class is either represented structurally or belongs to a certified non-reaching ledger generator, which is restored exhaustiveness relative to $\Sigma$.

If the quotient is nonzero, choose a nonzero represented or compactified quotient class. Target-completeness excludes zero target capacity and zero target flux for such a class; otherwise it would lie in the closed ledger and vanish in the quotient. Hence it is a positive-capacity, nonzero-flux target-visible boundary-defect class. The declared interface is exhaustive after quotienting by the closed ledger: any further proposed class is either in the ledger or represented by this quotient. The realization/nonpolarity certificate turns positive target-visible-channel capacity into an admissible reaching path. $\square$

### Theorem 4.7 (Fast-Track Equals the Explicit Boundary-Defect Calculation)
The boundary-defect fast-track gives the same verdict as the non-fast-track repair that explicitly stratifies every boundary defect into rigid, cost-divergent, residual, gauge, dyadic, rebased, subthreshold, separator-killed, and zero-capacity pieces.

More precisely, an explicit repair transcript proves that no boundary defect reaches $\Sigma$ if and only if it proves $\mathfrak H_{\rm bdry}^{\Sigma}=0$ for the target-complete ledger generated by its pieces. Conversely, a proof that $\mathfrak H_{\rm bdry}^{\Sigma}=0$ induces such a transcript by quotient membership: every boundary-defect class has a representative in a named ledger generator, or is the represented limit of such generators in the closed ledger topology.

**Proof.** Given an explicit repair transcript, place each certified non-reaching stratum into $\mathcal K_{\rm bdry,closed}^{\Sigma}$. The transcript's exhaustiveness assertion says these strata cover $\widehat{\mathcal H}_{\rm bdry}^{\mathcal B,\Sigma}$ after compactification. Since the ledger is closed, the quotient by the ledger is zero, so $\mathfrak H_{\rm bdry}^{\Sigma}=0$.

Conversely, if $\mathfrak H_{\rm bdry}^{\Sigma}=0$, every compactified boundary-defect element lies in the closed ledger. By Definition 4.5, the ledger is generated only by named certified mechanisms: represented gradients, residual zero-flux classes, gauge/symmetry/lift readouts, dyadic decay, rebasing, subthreshold, rigidity, Pohozaev/Noether/virial/frequency separators, and zero-capacity classes. Choosing the corresponding generator or limiting generator for each element gives the explicit repair classification. The reachability conclusion is identical in both arguments because both use the same target-flux functional, the same capacity, and the same realization/nonpolarity rule for nonzero target-visible quotient classes. $\square$

### Remark 4.8 (What the Fast-Track Avoids)
The non-fast-track proof enumerates boundary-defect cases one stratum at a time and then proves the list is exhaustive. The fast-track computes the single quotient $\mathfrak H_{\rm bdry}^{\Sigma}$. Closed range is the special case $\overline{{\rm im}\,G}={\rm im}\,G$; then Theorem 4.2 puts every directed class in the represented structural-gradient/residual split, so the boundary-defect quotient has zero numerator after the ledger projection. In gapless settings, the page-count bottleneck is the target-complete ledger and the quotient computation, not a separate case-by-case calculation.


## 5. Committors, Quotients, and Operator Learning

### Definition 5.1 (Continuum Committor)
Let $L$ be a budget-admissible generator and let $A,B\subset X$ be disjoint represented boundary or terminal sets. The committor $h$ is the solution of
$$
Lh=0\quad\text{on }X\setminus(A\cup B),
\qquad h|_A=1,
\qquad h|_B=0,
$$
in the form sense. Probabilistically, $h(x)$ is the hitting probability of $A$ before $B$ for the associated right process, outside exceptional sets.

### Theorem 5.2 (Committor Quotient)
The level-set quotient
$$
q_h:X\to[0,1],\qquad q_h(x)=h(x),
$$
is the coarsest represented quotient, among quotients measurable with respect to $h$, that preserves the hitting verdict between $A$ and $B$. Any observable constant on the level sets of $h$ cannot distinguish states with the same boundary-hitting law.

**Proof.** The Markov property gives $h(x)=\mathbb P_x(\tau_A<\tau_B)$. If $h(x)=h(y)$, every boundary verdict measurable only through this hitting probability agrees at $x$ and $y$. Conversely, if two states have different $h$ values, the event “hit $A$ before $B$” separates them. Hence the quotient by $h$ is exactly the represented hitting-law quotient generated by the committor. $\square$

### Theorem 5.3 (Self-Defect/Harmonic Residual Identity)
For an approximate committor $u$, the residual $Lu$ with boundary mismatch is the lumpability/self-defect of the quotient generated by $u$. Its squared dual norm in the defect geometry is the effective-resistance certificate
$$
\|Lu\|_{\mathcal L_W^+}^2=\langle Lu,\mathcal L_W^+Lu\rangle,
$$
whenever the residual is routable.

**Proof.** Exact committors are exactly harmonic functions for the killed or boundary-conditioned generator. Therefore $Lu$ is the represented defect obstructing descent to the terminal two-point quotient. The compensation equation is the effective-resistance equation of Theorem 6.3 with $E=Lu$: the least-energy compensator is $\phi=\mathcal L_W^+Lu$ on the positive spectral subspace, and its energy is $\langle Lu,\mathcal L_W^+Lu\rangle$. $\square$

### Theorem 5.4 (Restricted Conditioning Controls Approximation)
Let $\mathcal R$ be a represented approximation class for committors. Assume the Galerkin restriction of $L$ to $\mathcal R\ominus\ker L$ is closed and coercive, with smallest active eigenvalue $\lambda_{\mathcal R}>0$, and assume the exact committor has an $L^2$ projection onto $\overline{\mathcal R}$. Then residual minimization in $\mathcal R$ is stable with inverse norm at most $\lambda_{\mathcal R}^{-1}$. The certified obstruction is the projection error of the exact committor outside $\overline{\mathcal R}$ plus the effective-resistance residual inside $\mathcal R$.

**Proof.** Residual minimization is a Galerkin projection of the harmonic equation onto $\mathcal R$. Coercivity on $\mathcal R\ominus\ker L$ gives the inverse bound by the spectral theorem or Lax-Milgram. Projection error gives the irreducible approximation component; the inverse applied to the residual gives the resistance component. $\square$


## 6. Defect-Aligned Geometry and Effective Resistance

### Definition 6.1 (Effective-Resistance Defect Energy)
Let $\mathcal L_W\ge0$ be self-adjoint. For a defect current $E$, define
$$
\mathcal R_W(E)=\|\mathcal L_W^{-1/2}P_{(\ker\mathcal L_W)^\perp}E\|^2,
$$
with value $+\infty$ if the displayed vector is not in $L^2$. In discrete or finite-rank form this is
$$
\mathcal R_W(E)=\sum_{\lambda_k>0}\lambda_k^{-1}|\langle E,v_k\rangle|^2.
$$

### Theorem 6.2 (Routable/Harmonic Split)
Every defect decomposes as
$$
E=E^{\rm rt}+E^{\rm harm},
\qquad E^{\rm harm}=P_{\ker\mathcal L_W}E,
\qquad E^{\rm rt}=P_{(\ker\mathcal L_W)^\perp}E.
$$
The routable part has finite certificate exactly when $\mathcal R_W(E^{\rm rt})<\infty$. The harmonic part has infinite resistance unless it vanishes.

**Proof.** The spectral theorem gives the orthogonal decomposition into kernel and positive spectral subspace. The pseudoinverse acts by $\lambda^{-1}$ on positive spectrum and cannot route a nonzero kernel component. Thus the positive spectral component is the only possibly routable one, and its cost is the stated spectral sum. $\square$

### Theorem 6.3 (Compensation Equation)
If $\mathcal R_W(E)<\infty$, then the equation
$$
\mathcal L_W\phi=E^{\rm rt}
$$
has a minimal-energy weak solution $\phi=\mathcal L_W^+E^{\rm rt}$ in the form domain, and
$$
\langle\phi,\mathcal L_W\phi\rangle=\mathcal R_W(E).
$$
This is the least possible energy among all compensators for the routable defect.

**Proof.** Functional calculus defines $\phi=\mathcal L_W^+E^{\rm rt}$ on the positive spectral subspace. For any other weak compensator $\psi$, $\psi-\phi\in\ker\mathcal L_W$, so the form energy is unchanged or larger after projecting away irrelevant null components. The spectral identity gives the displayed energy. $\square$

### Theorem 6.4 (Defect-Aligned Drift Bound)
Suppose the antisymmetric contribution of the generator pairs with observables through a routable defect $E$ and symmetric form $\mathcal E^{\mathsf s}(u,u)=\langle u,\mathcal L_Wu\rangle$. Then
$$
|\langle E,u\rangle|
\le \mathcal R_W(E)^{1/2}\,\mathcal E^{\mathsf s}(u,u)^{1/2}
+\|E^{\rm harm}\|\,\|u\|.
$$
If $E^{\rm harm}=0$, the whole defect is controlled by the defect-aligned symmetric energy.

**Proof.** Write $E^{\rm rt}=\mathcal L_W\phi$. Then
$$
|\langle E^{\rm rt},u\rangle|=|\langle\mathcal L_W^{1/2}\phi,\mathcal L_W^{1/2}u\rangle|
\le \mathcal R_W(E)^{1/2}\mathcal E^{\mathsf s}(u,u)^{1/2}.
$$
The harmonic term is bounded by Hilbert-space Cauchy-Schwarz. $\square$


### 6.4.1 Full Defect-Geometry Inequality Profile

Effective resistance is the Poincare face of the continuum defect geometry. The profile is not an ambient regularity assumption: \(\mathcal L_W\) is the represented form whose purpose is to route or reject the active defect \(E\). When the same form carries additional inequalities, the routing theorem can use

$$
\Pi_D=(\mathcal L_W,E,\rho_D,\lambda_D,\alpha_{\rm MLSI}^D,\kappa_D,\operatorname{Cap}_D,\Phi_D,\alpha_T^D).
$$

The selectable faces are used slot by slot. The routing slot contains

$$
\mathcal C_{{\rm route},{\rm Poincare}}=\mathcal R_W(E).
$$

The entropy slot may contain an MLSI concentration bound, with constants fixed by the declared form normalization:

$$
\mathcal C_{\rm MLSI}=c_{\rm MLSI}
\sqrt{\frac{\mathcal I_D(\rho_D)/\alpha_{\rm MLSI}^D+\tau_\delta}{m}}.
$$

The propagation slot may contain a BE contraction bound, for example

$$
\mathcal C_{\rm BE}=c_{\rm BE}e^{-2\kappa_Dt}
\quad\text{or}\quad
c_{\rm BE}/\kappa_D,
$$

provided \(\kappa_D>0\) and the declared semigroup calculus proves the contraction estimate. Target access, death avoidance, and boundary bottlenecks use capacity with the correct polarity: inverse capacity for target-reaching rates, zero capacity for polar death/boundary excision, and the declared barrier when capacity is positive. Deployment or law-shift slots may use

$$
\mathcal C_{\rm tr}=c_{\rm tr}\operatorname{Lip}_D(\ell)
\sqrt{\mathrm{KL}(\nu\Vert\mu)/\alpha_T^D}.
$$

Any continuum, learning, or RL theorem that previously used only a Poincare propagation constant may replace the corresponding slot by

$$
\min_{j\in \mathcal J_s}\mathcal C_{s,j}(\Pi_D)+\tau_s,
$$

provided every candidate in \(\mathcal J_s\) bounds the same slot in the same units and the transcript records which theorem supplies that bound. Capacity remains the correct face for polar sets and boundary bottlenecks: zero capacity invokes Theorem 7.2, while positive capacity requires the realization/nonpolarity transcript of the relevant target-visible class. Bakry--Emery is the active stability face: the declared semigroup calculus must prove contraction of defect gradients, and only then can a Caus repair direction be barred from amplifying local inconsistency. MLSI is the entropy face: the transcript must prove dissipation of the represented defect density through \(\mathcal L_W\). In exact continuum settings a BE theorem may imply the entropy or transport face; in Galerkin, finite-element, or sampled settings, MLSI remains a separate witness unless that implication is proved on the declared approximation class. If no represented local geometry supplies the routing, entropy, or curvature-stability certificate for the active defect, the row returns a nonzero quotient/profile obstruction rather than a smoothed closure.



### 6.5 Commutator Routing Fast-Track for Degenerate Geom

When $\mathcal L_W$ is degenerate, direct effective resistance may be infinite and the routing object is no longer $\mathcal L_W^+$ alone. The correct object is the commutator $[\mathsf{Caus},\mathsf{Geom}]$ modulo the Noether gauge algebra. Two fast-tracks then appear: a hypocoercive bracket-rank route when $\mathcal L_W\neq0$, and a Morawetz positive-commutator route when $\mathcal L_W=0$.

### Definition 6.5 (Degenerate Geom and the Caus generator)
Let the generator split on a represented common invariant core $\mathcal D$ as
$$
L=-\mathcal L_W+\mathsf A,
\qquad \mathcal L_W=\mathcal L_W^*\ge0,
\qquad \mathsf A^*=-\mathsf A,
$$
with $\mathcal L_W$ the symmetric Geom channel and $\mathsf A=L^{\mathsf{skew}}$ the antisymmetric Caus channel. The commutators below are required to be represented on $\mathcal D$, closable, and continuously extend to the declared form/Galerkin topology. In finite-dimensional or finite-element certificates these properties are verified by the supplied matrices/forms and their common domain. In continuum certificates they are explicit interface hypotheses. The coercive hypothesis is not assumed. The routing hypothesis is replaced by bracket or commutator closure below.

### Definition 6.6 (Noether gauge algebra)
Let $\mathfrak g$ be the symmetry algebra coming from the variational or conservation-law interface, with represented moment-map directions $\mathcal J(\xi_a)$. The gauge subspace is
$$
\mathcal G=\operatorname{span}\{\text{conserved directions}\}=\ker(\mathcal L_W)\cap\{\mathsf A\text{-invariant directions}\}.
$$
These directions are undamped by design and are quotient out before any rank or sign test.

### Definition 6.7 (Caus-commutator tower)
The bracket tower of Geom under Caus is
$$
\mathcal L_W^{(0)}=\mathcal L_W,\qquad \mathcal L_W^{(k+1)}=[\mathsf A,\mathcal L_W^{(k)}],
$$
where each commutator is interpreted as the represented closure of its action on $\mathcal D$. The positive bracket-augmented closure is
$$
\overline{\mathcal L_W}^{\,\mathrm{ad},K}=\sum_{k=0}^{K}c_k\,(\mathcal L_W^{(k)})^*\mathcal L_W^{(k)},
\qquad c_k>0.
$$
This positive operator, not the raw skew commutator itself, is the routing geometry. It is the operator form of the bracket-rank / twisted-energy construction.

### Theorem 6.8 (Certified Hörmander Span = Symplectic Rank Full Mod Gauge = Finite Hypocoercive Resistance)
On $\mathcal G^\perp$, the hypocoercive routing certificate consists of the following equivalent quantitative statements:

1. there exists $K$ such that the closed span of $\sum_{k=0}^{K}\operatorname{im}\mathcal L_W^{(k)}$ is $\mathcal G^\perp$ and the associated bracket Gramian has a positive lower bound;
2. the Caus flow has no non-gauge symplectic-invariant subspace inside $\ker\mathcal L_W\cap\mathcal G^\perp$, and the represented closed-range/gap certificate for the bracket tower is supplied;
3. $\overline{\mathcal L_W}^{\,\mathrm{ad},K}\ge\gamma_{\mathrm{hyp}}\,\mathrm{Id}$ on $\mathcal G^\perp$ for some $\gamma_{\mathrm{hyp}}>0$;
4. the bracket-augmented resistance
$$
\mathcal R_W^{\mathrm{hyp},K}(E)=\langle E,(\overline{\mathcal L_W}^{\,\mathrm{ad},K})^+E\rangle
$$
is finite for every $E\in\mathcal G^\perp$ and satisfies $\mathcal R_W^{\mathrm{hyp},K}(E)\le\gamma_{\mathrm{hyp}}^{-1}\|E\|^2$.

**Proof.** (1)$\Leftrightarrow$(3): the bracket Gramian associated with the tower is $\overline{\mathcal L_W}^{\,\mathrm{ad},K}$. The certified closed-span condition says precisely that this nonnegative self-adjoint operator has no quotient kernel and has a lower spectral bound $\gamma_{\mathrm{hyp}}>0$ on $\mathcal G^\perp$. The lower bound is the closed-range part of the certificate; algebraic density alone is not used. (1)$\Leftrightarrow$(2): a non-gauge $\mathsf A$-invariant subspace in $\ker\mathcal L_W$ is missed by every bracket exactly when the bracket Gramian has a nonzero quotient kernel. Removing such subspaces modulo gauge is the symplectic-rank condition, with the same closed-range/gap certificate. (3)$\Leftrightarrow$(4): coercivity of the self-adjoint closure on $\mathcal G^\perp$ is equivalent to boundedness of its pseudoinverse there, and the displayed resistance bound is the spectral theorem. $\square$

### Theorem 6.9 (Hypocoercive carrier decreases after bracket mixing)
If the bracket tower has the certified gap of Theorem 6.8 at depth $K$, there exists a twisted carrier $\mathcal K_{\mathrm{tw}}$ equivalent to $\|u\|^2$ on $\mathcal G^\perp$ such that
$$
\frac{d}{d\tau}\mathcal K_{\mathrm{tw}}\le-\gamma_{\mathrm{hyp}}\,\mathcal K_{\mathrm{tw}}
$$
when a spectral gap is present, or an algebraic decay inequality with rate controlled by $K$ in the degenerate-gap case.

**Proof.** The cross terms are chosen so that differentiating $\mathcal K_{\mathrm{tw}}$ along the generator produces, besides the direct dissipation, the bracket terms up to depth $K$. Their sum is controlled by the coercive closure of Theorem 6.8 on $\mathcal G^\perp$. Equivalence of $\mathcal K_{\mathrm{tw}}$ with $\|u\|^2$ follows by taking the cross-term coefficients small. With the spectral gap $\gamma_{\mathrm{hyp}}$, integration gives the displayed exponential decay. If the interface supplies only a subcoercive bracket estimate of order $K$, the same differentiated carrier gives the corresponding algebraic inequality with the declared subcoercive exponent. $\square$

### Definition 6.10 (Hypocoercive routing invariant)
The fast-track invariant is the bracket-rank deficit modulo gauge, together with the quantitative gap certificate of Theorem 6.8,
$$
\mathfrak h^\Sigma=\mathcal G^\perp\Big/\overline{\sum_{k=0}^{K}\operatorname{im}\mathcal L_W^{(k)}}.
$$
Vanishing of $\mathfrak h^\Sigma$ plus the lower bound $\gamma_{\mathrm{hyp}}>0$ is equivalent to hypocoercive routing.

### Theorem 6.11 (Hypocoercive fast-track dichotomy)
Assume the hypocoercive row supplies the represented bracket tower and the closed-span/gap transcript of Theorem 6.8. Exactly one alternative holds.

- If $\mathfrak h^\Sigma=0$ and the gap bound $\gamma_{\mathrm{hyp}}>0$ is certified, the degenerate Geom channel routes the defect through the bracket closure, the twisted carrier decreases, and the defect certificate is finite on $\mathcal G^\perp$.
- If $\mathfrak h^\Sigma\ne0$, the deficient subspace is a non-gauge $\mathsf A$-invariant symplectic subspace in $\ker\mathcal L_W\cap\mathcal G^\perp$; it is a conserved Caus current, and it reaches the target if and only if its affordable target flux is nonzero.

**Proof.** The non-reaching branch is Theorem 6.8 plus Theorem 6.9. The target-reaching alternative is the contrapositive of Theorem 6.8(2): a missing bracket direction is preserved by the Caus flow and carries no dissipation, hence it is a persistent current. The target-flux classification is Theorem 13.7. $\square$

### Corollary 6.12 (Coverage: kinetic and partially dissipative PDE)

In the represented finite-depth certificates the bracket depth is one or two, and the gauge is given by the collision invariants.

**Proof.** In such a transcript, the commutator identities exhibit the transport-bracket directions generated from the dissipative part, and the supplied bracket Gramian lower bound verifies the certified gap on the quotient after collision invariants are removed. Theorem 6.11 then gives the routing verdict. $\square$

### Definition 6.13 (Conjugate operator and Mourre commutator)
Now assume $\mathcal L_W=0$ and write the conservative generator as $L=\mathsf A=-iH$. A conjugate operator is a self-adjoint $A$ such that the Mourre commutator
$$
i[H,A]=:B
$$
is sign-definite modulo gauge on the relevant spectral subspace $P=\mathbf 1_\Delta(H)$, meaning $PBP\ge\theta P$ for some $\theta>0$ after the gauge projection has been removed.

### Theorem 6.14 (Positive commutator gives a monotone carrier)
If a conjugate operator $A$ with Mourre constant $\theta>0$ exists modulo gauge, then the expectation $\langle A\rangle_u(\tau)=\langle u(\tau),Au(\tau)\rangle$ is monotone and
$$
\frac{d}{d\tau}\langle A\rangle_u=\langle u,Bu\rangle\ge\theta\,\|Pu\|^2.
$$
Hence the Morawetz carrier yields an integrated spacetime estimate even though there is no dissipation in the generator.

**Proof.** Heisenberg’s identity gives $\frac{d}{d\tau}\langle A\rangle_u=\langle u,i[H,A]u\rangle$. The Mourre lower bound on the positive spectral subspace gives the displayed inequality after gauge projection. Integrating in time yields the monotone-carrier estimate. $\square$

### Definition 6.15 (Morawetz routing invariant)
The fast-track invariant is the Mourre deficit modulo gauge: the spectral subspace on which no conjugate operator produces a positive commutator. Vanishing of this deficit is equivalent to a global Morawetz carrier modulo gauge.

### Theorem 6.16 (Morawetz fast-track dichotomy)
Exactly one alternative holds.

- If the Morawetz deficit vanishes, the conservative system has a monotone carrier, no concentration occurs on the target shell, and the routing certificate is complete.
- If the Morawetz deficit is nonzero, the represented spectral/compactified quotient contains a non-gauge class on which the commutator has no positive sign, such as an embedded resonance, standing wave, or soliton in the corresponding PDE transcript. It reaches the target if and only if its affordable target flux is nonzero.

**Proof.** The non-reaching branch is Theorem 6.14. The target-reaching alternative is the represented quotient on which positivity fails after gauge projection. In concrete dispersive transcripts this quotient is represented by the named resonance, standing-wave, soliton, or compactified spectral-defect class supplied by the problem interface. The affordable-flux classification is Theorem 13.7. $\square$

### Corollary 6.17 (Coverage: dispersive PDE)
Nonlinear Schrödinger, KdV, and wave-map flows in conservative form are covered by the Morawetz fast-track when their represented virial/Morawetz transcript supplies a positive commutator modulo the conserved Noether quantities. The conjugate operator is the dilation or weighted virial multiplier, and the gauge consists of conserved quantities such as mass, momentum, and energy, plus any scaling or Galilean symmetry directions.

**Proof.** In each supplied transcript, the virial/Morawetz identity is a positive-commutator statement modulo conserved quantities. Theorem 6.16 applies to that transcript, and any soliton, standing-wave, or spectral-defect quotient left by the commutator row is the positive-flux target-reaching alternative. $\square$

### Theorem 6.18 (The commutator $[\mathsf{Caus},\mathsf{Geom}]$ is the routing object when Geom is degenerate)
Let $\mathcal L_W\ge0$ be degenerate. The defect routes through $[\mathsf A,\mathcal L_W]$ modulo the Noether gauge algebra in exactly one of two modes:

2. **Sign mode, dispersive.** If $\mathcal L_W=0$, routing succeeds iff a conjugate operator makes $i[H,A]\succeq0$ modulo gauge.

In both cases the gauge directions are quotiented out before the rank or sign test. The two invariants are the closed-span/gap cokernel of the bracket tower and the positive-commutator deficit.

**Proof.** When $\mathcal L_W\ne0$, the relevant question is whether the bracket tower has the closed-span/gap certificate of Theorem 6.8. When $\mathcal L_W=0$, the tower is empty and the only remaining object is the conjugate commutator, which is Theorem 6.14. In both cases the gauge directions are invariant and dissipation-free, so they are removed before the test. $\square$

### Theorem 6.19 (Symplectic-rank reading)
Both fast-tracks are conservation of symplectic structure modulo gauge. Routing succeeds iff the symplectic flow has no non-gauge invariant quotient detected by the routing object. For the hypocoercive case this means no non-gauge invariant subspace in $\ker\mathcal L_W\cap\mathcal G^\perp$ after the closed-span/gap certificate is applied; for the dispersive case it means no non-gauge invariant subspace in $\mathcal G^\perp$ on which the virial commutator vanishes.

**Proof.** This is the symplectic reformulation of Theorems 6.8 and 6.16. A missing bracket direction or a missing positive commutator is exactly an invariant conserved current disjoint from the routing object and not gauge. $\square$

### Corollary 6.20 (Unified routing fast-track)
The framework’s routing certificate extends to coercive, hypocoercive, and dispersive Geom by replacing $\mathcal L_W^+$ with the appropriate positive routing object:

\mathcal R_W^{\mathrm{rout}}(E)=
\begin{cases}

This is finite modulo gauge iff the certified closed-span/gap condition holds in the hypocoercive case or the Mourre constant is positive in the dispersive case.

**Proof.** The coercive case is the depth-zero instance. The hypocoercive case is Theorem 6.8, and the dispersive case is Theorem 6.14. Since the gauge is quotiented out in all cases, the same routing verdict is returned by the corresponding fast-track. $\square$

## 7. Capacity, Polarity, and Unreachability

### Definition 7.1 (Polar Target)
A target $\Sigma$ is polar for a quasi-regular form if the associated right process hits it with probability zero for quasi-every starting point. In the budgeted framework this is read through ${\rm Cap}_{\mathcal B}(\Sigma)=0$.

### Theorem 7.2 (Zero Capacity Implies No Affordable Path)
If ${\rm Cap}_{\mathcal B}(\Sigma)=0$, then $\Sigma$ is polar for the affordable form: for quasi-every $x$,
$$
\mathbb P_x(\exists t\ge0:X_t\in\Sigma)=0.
$$
No finite-energy affordable observable can concentrate nontrivially on $\Sigma$.

**Proof.** For quasi-regular forms, capacity-zero sets are exceptional: they are outside an increasing nest of compact sets on which the process lives. The associated right process therefore avoids them outside a properly exceptional set. Since the equilibrium-potential energy defining capacity is zero, any affordable finite-energy function is determined up to quasi-continuous modification off $\Sigma$. $\square$

### Theorem 7.3 (Quotient Pullback of Polarity)
Let $q:X\to Q$ be represented and terminal-lumpable for $\Sigma$, meaning $\Sigma=q^{-1}(q(\Sigma))$ up to null sets. Suppose that for every decreasing represented neighborhood basis $V_r\downarrow q(\Sigma)$ with quotient equilibrium potentials $f_r$, there are affordable compensators $\psi_r$ on $X$ such that $\psi_r\ge1$ on $q^{-1}(V_r)$ and
$$
\mathcal E_X(\psi_r,\psi_r)
\le C\,\mathcal E_Q(f_r,f_r)+\varepsilon(r),
\qquad \varepsilon(r)\to0.
$$
Then ${\rm Cap}_{\mathcal B}^Q(q(\Sigma))=0$ implies ${\rm Cap}_{\mathcal B}^X(\Sigma)=0$. More generally,
$$
{\rm Cap}_{\mathcal B}^X(\Sigma)
\le C\,{\rm Cap}_{\mathcal B}^Q(q(\Sigma))+\liminf_{r\downarrow0}\varepsilon(r).
$$
A finite nonvanishing defect cost gives only this quantitative upper bound; polarity requires the vanishing local correction above.

**Proof.** The functions $\psi_r$ are admissible competitors for the capacity of $\Sigma$ because they dominate $1$ on neighborhoods of $\Sigma$. Taking the infimum over represented neighborhood bases and using the displayed energy inequality gives the quantitative bound. If the quotient capacity is zero and $\varepsilon(r)\to0$, the right-hand side tends to zero. Therefore ${\rm Cap}_{\mathcal B}^X(\Sigma)=0$, and Theorem 7.2 gives polarity on $X$. $\square$

### Theorem 7.4 (Symmetrizing Gauge Criterion)
If a directed drift becomes a gradient drift $-\nabla\Psi$ after a represented quotient, then the operator
$$
L_\Psi=\Delta-\nabla\Psi\cdot\nabla
$$
is symmetric in $L^2(e^{-\Psi}d\mu)$ with form
$$
\mathcal E^\Psi(f,f)=\int|\nabla f|^2e^{-\Psi}d\mu.
$$
The gauge is admissible exactly when the density $e^{-\Psi}$ and its reciprocal are bounded by budget factors on the represented affordable region.

**Proof.** Integration by parts gives
$$
\int L_\Psi f\,g\,e^{-\Psi}d\mu=-\int\nabla f\cdot\nabla g\,e^{-\Psi}d\mu,
$$
which is symmetric and nonnegative. The Radon-Nikodym density condition is precisely the admissible-presentation condition for changing the reference measure using only declared presentation data rather than a solution-dependent gauge. $\square$


## 8. Dissipation-Budget Reaching Dichotomy

This section states the reaching theorem as a conditional metatheorem. It applies after the abstract channel, scaling, and budget certificates listed below are supplied.

### Definition 8.1 (Scaling-Compatible Reaching Setup)
Assume a renormalization map $\mathcal T$ acts on the represented state space or quotient and preserves the channel decomposition. A reaching descent to $\Sigma$ is represented by iterates of $\mathcal T$ over infinitely many log-scale steps. A Caus current is scale-fixed if it lies in $\ker(\mathcal T-I)$ in the structural-gradient space. The scale-compatible interface also supplies a certified splitting
$$
\mathcal S_{\rm Caus}=\mathcal S_{\rm fix}\oplus\mathcal S_{\rm mov},
\qquad \mathcal S_{\rm fix}=\ker(\mathcal T-I),
$$
and a scale-stability certificate: every reaching Caus current has zero $\mathcal S_{\rm mov}$ projection, equivalently the moving projection has zero equilibrium flux through shrinking target shells or is reclassified as residual by a verified irreducibility certificate.

### Theorem 8.2 (Scale-Stable Reaching is Fixed Caus)
Assume continuum exhaustiveness, zero flux of residual currents through target shells, preservation of the structural-gradient space by $\mathcal T$, the certified splitting of Definition 8.1, and the scale-stability certificate. Then any directed current reaching $\Sigma$ has its reaching component in the scale-fixed Caus subspace $\ker(\mathcal T-I)$. The moving component either vanishes for reaching or belongs to the residual zero-flux class by certificate.

**Proof.** Exhaustiveness decomposes any directed current into structural-gradient and residual components by Theorem 4.2 or by the vanished boundary-defect quotient of Theorem 4.6. Residual currents with zero target flux are placed in $J_{\rm res}^{0\Sigma}$ by Definition 4.5 and Lemma 13.6, so Theorem 7.2 and target-completeness exclude them from the affordable reaching channel. The structural-gradient component decomposes as $F_{\rm fix}+F_{\rm mov}$ under the certified splitting. The scale-stability certificate states that $F_{\rm mov}$ either has zero reaching flux, hence enters the same residual ledger, or is represented by the certificate as a barred residual class. Therefore the only remaining target-visible reaching component is $F_{\rm fix}\in\ker(\mathcal T-I)$. $\square$

### Theorem 8.3 (Finite Dissipation Excludes Positive Per-Scale Fixed Points)
Suppose every nonzero scale-fixed Caus profile reaching $\Sigma$ carries a positive dissipation quantum $c_*>0$ per log-scale step, and suppose the budget charges accumulated dissipation. Then no such profile is affordable under any finite ceiling $b(n)$, because
$$
\sum_{j=0}^{\infty}c_*=+\infty.
$$
Consequently $\Sigma$ is unreached whenever all possible reaching currents must pass through such fixed Caus profiles.

**Proof.** By Definition 8.1, a reaching descent to $\Sigma$ is represented by infinitely many log-scale iterates. A scale-fixed profile has the same represented dissipation cost on each iterate, and the hypothesis gives the uniform lower bound $c_*>0$. B1-B4 make accumulated dissipation additive/subadditive in the declared budget ceiling, so the partial costs dominate $Nc_*$ on the first $N$ scale steps and diverge as $N\to\infty$. This contradicts any finite ceiling $b(n)$. If Theorem 8.2 exhausts all reaching channels into these charged fixed profiles plus residual classes in $J_{\rm res}^{0\Sigma}$, every affordable target-reaching channel is removed. $\square$

### Corollary 8.4 (Friction/Dissipation Dichotomy)
Under the hypotheses of Theorems 8.2 and 8.3, finite positive dissipation budget implies ${\rm Cap}_{\mathcal B}(\Sigma)=0$. If the budget does not charge the fixed Caus profile, this theorem gives no exclusion.

**Proof.** Theorem 8.3 places every charged scale-fixed reaching profile outside the finite budget, and Theorem 8.2 places the moving and residual pieces in zero-flux residual ledger classes. In the target-complete ledger, the remaining represented channels have zero affordable target flux or are unaffordable. The capacity envelope of Definition 2.2 therefore has no positive-energy admissible target witness, so the zero-capacity alternative of Theorem 7.2 applies. If the fixed profile is uncharged, the divergence argument in Theorem 8.3 is unavailable. $\square$

## 10. Stratified Cost-Rigidity Closure Engine

### Definition 10.1 (SCRC Interface)
A continuum problem supplies three abstract objects:

1. A concentration or defect threshold $\varepsilon_*$ such that subthreshold profiles are not boundary defects.
2. A nonnegative cost density $\mathfrak C(\tau)$ in log-scale time.
3. A rigidity functional $\mathcal J(\tau)$ detecting confinement of the renormalization orbit.

The cost must satisfy:

- **P1 local integrability:** $\int_{\tau_0}^{T}\mathfrak C<\infty$ for finite $T$.
- **P2 lower semicontinuity:** $\mathfrak C(\lim V_n)\le\liminf\mathfrak C(V_n)$ under profile convergence.
- **P3 carrier monotonicity:** a validated carrier decreases by at least the accumulated cost, up to explicitly budgeted boundary/constraint terms with constants uniform on nested finite windows.

The interface also supplies an in-window rigidity/cost certificate: for every represented in-window log-scale profile path $V(\tau)$, if $V$ remains above the defect threshold on a tail and is not eventually in the declared rigid class, then
$$
\int_{\tau_0}^{T}\mathfrak C(\tau)\,d\tau\to\infty
\qquad (T\to\infty).
$$
Escaping profile paths are excluded from this check by Lemma 13.6: their affordable target flux and affordable structural pairings vanish, so they are ledger generators in $J_{\rm res}^{0\Sigma}$. Thus every persistent in-window profile is forced either below threshold, into the rigid class, or into the cost-divergent class.

### Theorem 10.2 (No In-Window Residual Stratum)
Assume the SCRC interface, including the in-window rigidity/cost certificate. Then there is no in-window residual stratum: every persistent in-window boundary defect is either rigid or cost-divergent. Escaping boundary-defect classes are residual zero-flux ledger classes by Lemma 13.6 and are not passed to the SCRC quotient.

**Proof.** Let $V(\tau)$ be a persistent in-window boundary-defect profile. Persistence keeps $V$ above the threshold on a log-scale tail by Definition 10.1. The in-window rigidity/cost certificate is a two-way partition on such profiles: eventual membership in the declared rigid class gives the rigid stratum, while failure of eventual rigidity gives the displayed divergent accumulated cost and hence the cost-divergent stratum. Escaping profiles are outside the in-window hypothesis and have been placed in $J_{\rm res}^{0\Sigma}$ by Lemma 13.6. These three classes exhaust the represented compactified boundary-defect alternatives, so no target-visible in-window residual stratum remains. $\square$

### Theorem 10.3 (SCRC Master Exclusion)
Assume:

1. escaping classes removed to $J_{\rm res}^{0\Sigma}$ by Lemma 13.6 and no in-window residual stratum by Theorem 10.2;
2. cost-divergent stratum excluded by carrier monotonicity;
3. rigid stratum excluded or reclassified into a closed structural class by a declared rigidity certificate.

Then the target-visible boundary-defect quotient is empty and no target-reaching singular profile remains in the admissible closure.

**Proof.** Lemma 13.6 removes escaping classes to the residual zero-flux ledger. Theorem 10.2 reduces the in-window partition to the rigid and cost strata. The cost stratum is closed by the carrier monotonicity hypothesis: the accumulated cost diverges on every persistent representative, so the finite-budget condition excludes it by the same B1-B4 accumulation mechanism used in Theorem 8.3. The rigid stratum is either placed in the closed structural ledger by the declared rigidity map or excluded by its rigidity/Liouville certificate. These placements are precisely the generators of the target-complete boundary-defect ledger in Definition 4.5, so the quotient $\mathfrak H_{\rm bdry}^{\Sigma}$ vanishes by Theorem 4.6. $\square$

### Remark 10.4
SCRC delegates concrete verification to structural certificates: the threshold, P1-P3, the in-window rigidity/cost certificate, and the rigidity certificate. Escaping-profile closure is not one of those certificates; it is the unconditional affordable-flux statement of Lemma 13.6. Once the certificates are supplied, the stratification architecture need not be reproved.


## 11. Conservation Fast-Track

This section gives a general shortcut for verifying the SCRC cost hypotheses when the nonlinearity has a conservative structure. The conditional mechanism is: divergence form supplies a carrier identity, and exclusion follows only after a draining flux, passivity radius, separator, or other named sign certificate is supplied.

### Definition 11.1 (Criticality Deficit and Conservative Carrier)
For an evolution
$$
\partial_t u=-\mathcal A u+\mathcal N(u),
$$
let $\mathcal A$ have dissipative order $2s_{\rm diss}$ and let $s_{\rm crit}$ be the critical order of the scale-invariant norm. Define
$$
\kappa=s_{\rm diss}-s_{\rm crit}.
$$
Assume $\kappa\ge0$. Suppose
$$
\mathcal N(u)=\nabla\cdot F(u)+\mathcal Q(u),
$$
where $F$ is a represented flux and $\mathcal Q$ is a constraint response whose tail is budget-admissible. For a cutoff $\phi$, set
$$
\mathcal K(\tau)=\frac12\int\phi^2|V(\tau)|^2,
\qquad
\mathfrak C(\tau)=\int\phi^2|\nabla^{s_{\rm diss}}V(\tau)|^2.
$$

### Theorem 11.2 (Carrier Identity)
Under the above assumptions, the renormalized carrier satisfies
$$
\frac d{d\tau}\mathcal K(\tau)
= -\mathfrak C(\tau)+\Phi(\tau)+\mathcal P(\tau),
$$
where $\Phi$ is supported on the cutoff boundary and is determined by the flux $F$, and $\mathcal P$ is the constraint response.

**Proof.** Multiply the renormalized equation by $\phi^2V$ and integrate. The symmetric dissipative part gives $-\mathfrak C$ plus cutoff terms. The divergence term is integrated by parts, so its interior contribution becomes a boundary flux supported where $\nabla\phi\ne0$. The remaining nonlocal or constraint contribution is $\mathcal P$. $\square$

### Theorem 11.3 (Fast-Track Conditions for SCRC)
Assume $\kappa\ge0$, finite-window local energy estimates for the dissipative form, local compactness sufficient for the Aubin-Lions step, and budget-admissibility of the constraint tail. Then P1 and P2 for $\mathfrak C$ hold. If, in addition, there are constants $\eta,\alpha,C_P,C_\Phi$ independent of the finite window $[\tau_0,T]$ such that
$$
\int_{\tau_0}^{T}\mathcal P\,d\tau\le \eta\int_{\tau_0}^{T}\mathfrak C\,d\tau+C_P,
\qquad \eta<1,
$$
and the flux is draining in the sense that
$$
\int_{\tau_0}^{T}\Phi\,d\tau\le \alpha\int_{\tau_0}^{T}\mathfrak C\,d\tau+C_\Phi,
\qquad \alpha+\eta<1,
$$
then the carrier is monotone with a positive fraction of $\mathfrak C$ as its dissipation rate, so P3 holds with window-independent remainder $C_P+C_\Phi$.

**Proof.** P1 follows from finite-window local energy estimates for the dissipative form. P2 follows from compactness: bounded cost gives weak compactness at the dissipative order, the equation gives a time-derivative bound in a negative space, and Aubin-Lions gives strong convergence below the dissipative order; the squared norm is weakly lower-semicontinuous. Integrating Theorem 11.2 and using the two domination inequalities leaves
$$
\mathcal K(T)+(1-\alpha-\eta)\int_{\tau_0}^{T}\mathfrak C\,d\tau
\le \mathcal K(\tau_0)+C_P+C_\Phi,
$$
which is P3 with a uniform finite-window constant. $\square$

### Theorem 11.4 (Flux-Sign Exclusion Certificate)
If the SCRC threshold says every persistent nontrivial boundary defect carries nonzero cost over infinitely many log-scale windows, and the fast-track inequalities of Theorem 11.3 hold with uniform constants over the nested windows of the descent, then no persistent boundary defect exists. If the flux is feeding so that $\alpha+\eta\ge1$ or no domination inequality is available, the fast-track gives no exclusion.

**Proof.** A persistent nontrivial defect forces accumulated cost on arbitrarily long unions of log-scale windows. Applying the P3 inequality to $[\tau_0,T]$ bounds this accumulated cost by the finite initial carrier plus the fixed constant $C_P+C_\Phi$, independently of $T$. Letting $T\to\infty$ contradicts persistent positive cost. If the flux is feeding, the inequality has no positive dissipation fraction left, so the carrier need not decrease and no contradiction follows. $\square$


### Definition 11.5 (Elliptic Constraint-Tail Interface)
Let the constraint response be represented by an elliptic equation

\mathcal C p=S(V),\qquad \mathcal Q(V)=B p,

on an affordable window, with core $U_0\Subset U$. After a cutoff split $p=p_{\rm loc}+h_{\rm tail}$, the tail interface requires:

2. A represented dyadic partition of the exterior with annular contributions $h_k$.
3. A homogeneous-mode split $\mathcal H_{\rm hom}(U_0)=\mathcal H_{\rm low}\oplus\mathcal H_{\rm dec}$, where $\mathcal H_{\rm low}$ is finite-dimensional.
4. Summable decaying estimates
   $$
   \|P_{\rm dec}h_k\|_{X(U_0)}\le a_k\|\theta_kS(V)\|_{Y(A_k)},
   \qquad \sum_k a_k\|\theta_kS(V)\|_{Y(A_k)}\preceq b(n).
   $$
5. A represented gauge map from declared symmetries to low homogeneous modes,
   $$
   \Gamma:\mathfrak g\to\mathcal H_{\rm low}.
   $$

The tail obstruction is
$$
\mathfrak o_{\rm tail}=[P_{\rm low}h_{\rm tail}]\in\operatorname{coker}\Gamma.
$$

### Theorem 11.6 (Homogeneous Tails are Residual Source Channels)
If $\mathcal C h_{\rm tail}=0$ on the core, then for every compactly supported core test function $\varphi$,
$$
\langle h_{\rm tail},\mathcal C^*\varphi\rangle=0.
$$
Thus the tail has no local source in the core. In the channel taxonomy it is a residual constraint channel: it can contribute only through boundary traces, dyadic decay, or nondecaying gauge/cokernel modes.

**Proof.** The weak equation gives $\langle\mathcal C h_{\rm tail},\varphi\rangle=0$ for compactly supported $\varphi$. Moving $\mathcal C$ to the adjoint gives the displayed identity. The local source map therefore annihilates the tail on the core. $\square$

### Theorem 11.7 (Dyadic-Spatial Tail Closure iff Gauge Cokernel Vanishes)
Assume the elliptic constraint-tail interface of Definition 11.5. Then
$$
h_{\rm tail}=\Gamma\xi+h_{\rm dec}+h_{\rm obs},
$$
where $\Gamma\xi\in\operatorname{im}\Gamma$, the decaying component $h_{\rm dec}$ has budget-bounded tail norm, and $h_{\rm obs}$ represents $\mathfrak o_{\rm tail}$. Relative to the declared mechanisms of dyadic decay and gauge absorption, the tail is budget-admissible if and only if the dyadic decaying series is summable and $\mathfrak o_{\rm tail}=0$. If $\mathfrak o_{\rm tail}\ne0$, the obstruction is finite-dimensional and computable; it must be closed by an additional named certificate before Theorem 11.3 applies.

**Proof.** The dyadic partition gives a local homogeneous expansion $h_{\rm tail}=\sum_kh_k$ on the core. The projected decaying series is affordable by the summability hypothesis in Definition 11.5(4), so it is a charged dyadic ledger generator. The remaining low component lies in finite-dimensional $\mathcal H_{\rm low}$, hence decomposes by linear algebra into an image part in $\operatorname{im}\Gamma$ plus a cokernel representative. The image part is removed by the declared gauge action of Definition 11.5(5); the decaying part is closed by the dyadic summability ledger; the cokernel representative is the only remaining class. $\square$

### Corollary 11.8 (Constraint-Tail Fast-Track)

Theorem 11.7 exhausts that homogeneous tail into gauge, decay, and cokernel. Vanishing cokernel plus summable decay is exactly budget-admissibility of the tail; a nonzero residual transcript item leaves the corresponding target-visible residual class. $\square$


### Definition 11.9 (Reduced Flux-Dissipation Interface)
After quotienting gauge modes, closing elliptic constraint tails, and removing residual zero-flux channels, let $\mathcal P_{\rm crit}$ be the represented critical profile or trace space. The reduced carrier identity has the form
$$
\frac d{d\tau}\mathcal K(\tau)
\le -\mathfrak C(V(\tau))+\Phi(V(\tau))+R(\tau),
$$
where $\mathfrak C\ge0$ is the represented dissipation density, $\Phi$ is the represented boundary-flux functional, and $R$ is a uniformly budgeted remainder. The positive flux numerical radius is
$$
\beta_*=
\sup_{\substack{V\in\mathcal P_{\rm crit}\\ \mathfrak C(V)>0}}
\frac{\Phi_+(V)}{\mathfrak C(V)},
\qquad \Phi_+(V)=\max\{\Phi(V),0\}.
$$
Profiles with $\mathfrak C(V)=0$ are allowed in the reduced space only when $\Phi_+(V)=0$; otherwise they are declared feeding null modes.

### Theorem 11.10 (Flux Numerical-Radius Certificate)
Assume the reduced flux-dissipation interface. If $\beta_*<1$, then every represented profile path satisfies
$$
\mathcal K(T)+(1-\beta_*)\int_{\tau_0}^{T}\mathfrak C(V(\tau))\,d\tau
\le
\mathcal K(\tau_0)+\int_{\tau_0}^{T}R(\tau)\,d\tau.
$$
Thus the flux is draining with margin $1-\beta_*$ whenever the remainder is uniformly bounded on nested windows.

If $\beta_*>1$, then the declared interface contains represented feeding profiles, so no sign-only exclusion certificate is valid without an additional restriction. If $\beta_*=1$, the remaining sign work is the saturated equality quotient generated by the extremizer class
$$
\mathcal E_{\rm flux}
=\{V\in\mathcal P_{\rm crit}:\mathfrak C(V)>0,
\ \Phi_+(V)=\mathfrak C(V)\},
$$
or by limits of normalized extremizing sequences in the certified compactification. This quotient is $\mathfrak H_{\rm sat}^{\Sigma}$ after the saturation ledger and target-complete closed ledger are applied.

**Proof.** By definition of $\beta_*$, $\Phi(V)\le\Phi_+(V)\le\beta_*\mathfrak C(V)$ for all represented profiles with positive dissipation, and the null-mode clause handles $\mathfrak C=0$. Substituting this bound into the reduced carrier identity and integrating gives the displayed inequality. If $\beta_*>1$, the definition of supremum gives a represented profile or compactified extremizing sequence whose positive flux exceeds dissipation. If $\beta_*=1$, normalize by $\mathfrak C(V)=1$; the remaining objects are exactly profiles or compactified limits with deficit $1-\Phi_+(V)$ tending to zero. They are not closed by this radius estimate alone, and are passed to the certified compactification and quotient of Definition 11.13. $\square$

### Theorem 11.11 (Quadratic Flux Operator)
Suppose $\mathfrak C(V)=\langle AV,V\rangle$ and $\Phi(V)=\langle BV,V\rangle$ on the reduced Hilbert space, with $A\ge0$, $B=B^*$, after the kernel/gauge quotient has been taken. If $B$ is $A$-form-bounded, then
$$
\beta_*=
\max\left\{\sup\sigma(A^{-1/2}BA^{-1/2}),0\right\},
$$
with the Friedrichs-form realization used in the unbounded case. Hence the sign certificate is a spectral or finite-matrix computation whenever the reduced forms are discretely represented.

**Proof.** The $A$-form inner product turns the reduced domain into a Hilbert space. The symmetric flux form is represented by a self-adjoint bounded operator $S_{\rm flux}$ in that Hilbert space. Rayleigh-Ritz gives $\sup \langle BV,V\rangle/\langle AV,V\rangle=\sup\sigma(S_{\rm flux})$, and transporting by $A^{1/2}$ gives the displayed operator. Taking the positive part gives $\beta_*$. $\square$

### Theorem 11.12 (Quenched and Annealed Flux Certificates)
For stochastic problems, after the admissible Wiener lift, define
$$
\beta_*^{\rm q}=\operatorname*{ess\,sup}_\omega
\sup_{V\in\mathcal P_{\rm crit}^\omega}
\frac{(\Phi^\omega)_+(V)}{\mathfrak C^\omega(V)},
\qquad
\beta_*^{\rm ann}=
\sup_V
\frac{\mathbb E[(\Phi^\omega)_+(V)]}{\mathbb E[\mathfrak C^\omega(V)]}.
$$
Martingale terms are excluded from $\Phi^\omega$ and require their own quadratic-variation or concentration certificate. If $\beta_*^{\rm q}<1$, the deterministic draining inequality holds for almost every noise realization. If $\beta_*^{\rm ann}<1$, the corresponding annealed carrier inequality holds in expectation. Equality reduces to the quenched or annealed saturated target-visible-channel quotient.

**Proof.** The quenched claim applies Theorem 11.10 outside the null set in the essential supremum. The annealed claim follows from $\mathbb E[\Phi^\omega]\le\mathbb E[(\Phi^\omega)_+]\le\beta_*^{\rm ann}\mathbb E[\mathfrak C^\omega]$ and the expected carrier identity. In the equality case, normalize by the quenched or annealed dissipation and form the corresponding compactified extremizer class; its closure is then the quotient mechanism of Definition 11.13 and Theorem 11.14, with the probability mode fixed by the transcript. $\square$


### Definition 11.13 (Sharp Flux-Extremizer Obstruction)

where $Z$ may be an actual profile or a compactified limit of profiles with $\Phi_+(V_j)\to1$.

Let $\mathcal H_{\rm closed}$ be the declared closed harmless class generated by:

In nonlinear settings this is a quotient by the declared equivalence relation generated by $\mathcal H_{\rm closed}$. In Hilbert/quadratic settings it is the vector-space quotient of the top eigenspace or approximate top eigenspace.

### Theorem 11.14 (Borderline Flux Closes iff the Extremizer Obstruction Vanishes)
Assume $\beta_*=1$, the certified compactification of Definition 11.13, and lower semicontinuity of both $\mathfrak C$ and $\Phi_+$ on normalized extremizing sequences. If
$$
\mathcal O_{\rm ext}=0,
$$
then the borderline flux branch is closed: every normalized sequence with vanishing flux deficit
$$
\Delta_{\rm flux}(V_j)=\mathfrak C(V_j)-\Phi_+(V_j)\to0
$$
converges, after subsequence and declared equivalence, to a profile whose class lies in $\mathcal H_{\rm closed}$. Away from any certified neighborhood of $\widehat{\mathcal E}_{\rm flux}$ there is a positive margin
$$
\Phi_+(V)\le(1-\eta)\mathfrak C(V),\qquad \eta>0,
$$
so the strict draining theorem applies off the closed extremizer branch.

Conversely, if $\mathcal O_{\rm ext}\ne0$, the declared interface contains a critical neutral class with $\Phi_+=\mathfrak C$ that is not removed by gauge, rebasing, threshold, rigidity, or residual zero-flux closure. No theorem depending only on the previous certificates can exclude that class; the saturated quotient certificate of Definition 11.20 is required.

**Proof.** Normalize $\mathfrak C(V)=1$. Since $\beta_*=1$, the deficit $1-\Phi_+(V)$ is nonnegative. Compactness supplies subsequential limits for every deficit-vanishing sequence. Lower semicontinuity places every such limit in $\widehat{\mathcal E}_{\rm flux}$. If $\mathcal O_{\rm ext}=0$, each limit is equivalent to an element of $\mathcal H_{\rm closed}$; by the definition of $\mathcal H_{\rm closed}$ this means it is a gauge, rebased, subthreshold, rigid, or residual zero-flux ledger class. On the complement of a certified neighborhood of $\widehat{\mathcal E}_{\rm flux}$, compactness and nonnegativity of the deficit give a positive minimum $\eta$, which yields the strict bound. If $\mathcal O_{\rm ext}\ne0$, a representative of the nonzero quotient class has zero flux deficit and is not in any listed ledger generator. The target-complete saturated quotient of Definition 11.20 is therefore nonzero, so the equality row remains open until Theorem 11.17, 11.19, or 11.21 supplies a separator or target-visible alternative. $\square$

### Definition 11.16 (Saturation Remainder and Open-Channel Projection)
A proof of $\Phi_+\le\mathfrak C$ is represented by a finite saturation ledger

\Delta_{\rm flux}(V)=\mathfrak C(V)-\Phi_+(V)
=\sum_{a=1}^{N}D_a(V),
\qquad D_a(V)\ge0,

after all quotient, gauge, tail, and residual certificates have been applied. In the case $\beta_*=1$, an extremizer or extremizing limit satisfies $D_a=0$ for every ledger defect $D_a$.

The saturation equations $D_a=0$ define a represented saturation relation. Let
$$
\mathscr R_{\rm sat}:\widehat{\mathcal E}_{\rm flux}\to\mathcal K_{\rm chan}
$$
be the represented channel remainder determined by those equality equations: it is the component of the extremizing profile left after the inequalities used to prove the flux bound have forced all saturation defects $D_a$ to vanish. Decompose the channel space as
$$
\mathcal K_{\rm chan}
=\mathcal K_{\rm closed}\oplus\mathcal K_{\rm tv},
$$
where $\mathcal K_{\rm closed}$ is generated by certified channels: gauge/symmetry images, exact/Caus structural gradients with closed range, Geom dyadic-decay channels, rebased carriers, subthreshold classes, rigid closed classes, and the zero-flux residual class $J_{\rm res}^{0{\rm flux}}$. Let $\Pi_{\rm tv}$ denote projection to the quotient by $\mathcal K_{\rm closed}$.

The saturated target-visible-channel obstruction is
$$
\mathcal O_{\rm sat}
=
\overline{\Pi_{\rm tv}\mathscr R_{\rm sat}(\widehat{\mathcal E}_{\rm flux})}.
$$

### Theorem 11.17 (Saturation-to-Channel Closure)
Assume $\beta_*=1$ and the saturation ledger of Definition 11.16. If
$$
\mathcal O_{\rm sat}=0,
$$
then the borderline branch closes without a new rigidity calculation: every sharp flux extremizer or extremizing limit has its saturation remainder in $\mathcal K_{\rm closed}$, hence belongs to one of the closed ledger generators listed in Definition 11.16. In particular, if the saturation equations imply
$$
\mathscr R_{\rm sat}(Z)
\in
J_{\rm res}^{0{\rm flux}}+\mathcal H_{\rm closed}
\qquad\text{for every }Z\in\widehat{\mathcal E}_{\rm flux},
$$
then no further rigidity theorem is required for the flux sign.

If $\mathcal O_{\rm sat}\ne0$, the target-visible-channel quotient is the remaining equality row. This is strictly smaller than classifying all extremizers: projection by $\Pi_{\rm tv}$ removes every component in $J_{\rm res}^{0{\rm flux}}$ or another certified closed channel before the quotient is tested.

**Proof.** For $\beta_*=1$, any extremizing normalized sequence has $\Delta_{\rm flux}\to0$. Since the ledger defects are nonnegative and sum to $\Delta_{\rm flux}$, each $D_a\to0$ along the sequence. Passing to the certified compactified limit gives the saturation equations $D_a=0$. The map $\mathscr R_{\rm sat}$ records the remaining channel component after these equality constraints are imposed. If $\mathcal O_{\rm sat}=0$, the target-visible projection of every such remainder vanishes, so the remainder lies in $\mathcal K_{\rm closed}$; the relevant non-reaching mechanism is then the ledger generator containing that remainder. If the stronger displayed inclusion into $J_{\rm res}^{0{\rm flux}}+\mathcal H_{\rm closed}$ holds, the residual part has zero local production and zero target flux, while the rest is closed by definition. Conversely, a nonzero element of $\mathcal O_{\rm sat}$ is precisely a saturated equality remainder not removed by the channel ledger, so the separator or saturation row is necessary and sufficient for that quotient. $\square$

### Definition 11.18 (Pohozaev Separator Certificate)

\qquad Z\in\widehat{\mathcal E}_{\rm flux},

obtained from domain variations, scaling variations, multiplier identities, or stochastic averaged analogues after all boundary and martingale terms have been either represented in the carrier ledger or closed by previous certificates.

0=\sum_\ell c_\ell\mathcal P_\ell(Z)
=\|\Pi_{\rm tv}\mathscr R_{\rm sat}(Z)\|_{\mathcal Q}^{2}
+\sum_j q_j(Z)+b_{\rm closed}(Z),
$$
where $\|\cdot\|_{\mathcal Q}^{2}$ is a positive definite quadratic or convex certificate on the target-visible quotient, each $q_j\ge0$ is a represented saturation defect or subthreshold defect, and $b_{\rm closed}$ is a closed-channel ledger term whose value is zero after projection to $\mathcal K_{\rm chan}/\mathcal K_{\rm closed}$, equivalently a term cancelled by previously certified closed-channel identities.

### Theorem 11.19 (Pohozaev Separator Forces Vanishing Open Obstruction)
If a Pohozaev separator certificate exists, then
$$
\mathcal O_{\rm sat}=0.
$$
Hence the $\beta_*=1$ branch closes by Theorem 11.17. In the quadratic case, it is enough to certify that the Pohozaev quadratic form is strictly positive on
$$
\ker(I-S_{\rm flux})/\overline{\mathcal K}_{\rm closed}
$$
or on the corresponding approximate eigenspace/defect-measure quotient.

**Proof.** On $\widehat{\mathcal E}_{\rm flux}$, every Pohozaev identity equals zero. After quotienting by $\mathcal K_{\rm closed}$, the closed ledger term is zero by definition, so the separator writes zero as a sum of nonnegative represented terms on the quotient. Therefore each nonnegative term must vanish, in particular $\|\Pi_{\rm tv}\mathscr R_{\rm sat}(Z)\|_{\mathcal Q}^{2}=0$. Positive definiteness on the target-visible quotient gives $\Pi_{\rm tv}\mathscr R_{\rm sat}(Z)=0$ for every extremizer or extremizing limit. This is exactly $\mathcal O_{\rm sat}=0$. In the quadratic case, Corollary 11.15 identifies the equality branch with the top eigenspace or certified approximate eigenspace of $S_{\rm flux}$; restricting the positive separator to that quotient gives the displayed vanishing. $\square$

### Definition 11.20 (Target-Complete Closed Ledger)
Fix a target or blow-up set $\Sigma$ and a certified budget capacity ${\rm Cap}_{\mathcal B}$. Target-completeness means the converse also holds inside the declared compactified interface: if a represented or compactified saturated channel class has zero target capacity or zero certified normal target flux, then its class lies in $\mathcal K_{\rm closed}^{\Sigma}$. Define the saturated channel cohomology

### Theorem 11.21 (Exhaustive Saturated-Channel Dichotomy)
Assume a legal continuum interface, P1-P3, elliptic constraint-tail closure, gauge closure, dyadic/rebased-carrier closure, residual zero-flux closure, the flux-radius certificate with $\beta_*=1$, the compactified equality class $\widehat{\mathcal E}_{\rm flux}$, and a target-complete closed ledger $\mathcal K_{\rm closed}^{\Sigma}$. Then exactly one alternative holds.

**Closed alternative.** If
$$
\mathfrak H_{\rm sat}^{\Sigma}=0,
$$
then every saturated equality branch lies in a residual, gauge, exact/routed, dyadic, rebased, subthreshold, Pohozaev-killed, rigidly excluded, or zero-capacity ledger generator. The $\beta_*=1$ branch cannot reach $\Sigma$.

**Open alternative.** If
$$
\mathfrak H_{\rm sat}^{\Sigma}\ne0,
$$
then the declared compactified interface contains a represented or compactified saturated target-visible class with positive target capacity and nonzero certified target flux. If the interface also supplies a realization/nonpolarity certificate that this positive target-visible-channel capacity is attained by a budget-admissible path, then $\Sigma$ is reached. When $\Sigma$ is the blow-up set, this is the blow-up alternative.

There is no third alternative inside the certified interface. Any proposed additional remainder either has zero target capacity or zero target flux, hence lies in $\mathcal K_{\rm closed}^{\Sigma}$ by target-completeness, or has positive target capacity and is represented by the nonzero class in $\mathfrak H_{\rm sat}^{\Sigma}$.

**Proof.** The closed ledger is closed, so the quotient $\mathcal K_{\rm chan}/\mathcal K_{\rm closed}^{\Sigma}$ is a represented quotient and $\mathfrak H_{\rm sat}^{\Sigma}$ is well defined. If $\mathfrak H_{\rm sat}^{\Sigma}=0$, every represented or compactified saturation remainder lies in $\mathcal K_{\rm closed}^{\Sigma}$. Each generator of $\mathcal K_{\rm closed}^{\Sigma}$ is certified either to have zero target flux, to have zero target capacity, or to be excluded by a budget, rigidity, gauge, residual, dyadic, rebasing, Pohozaev, or Noether certificate. Closure preserves these zero-reach conclusions by lower semicontinuity of capacity and by continuity of the represented target-flux functional. Therefore the saturated branch cannot reach $\Sigma$.

If $\mathfrak H_{\rm sat}^{\Sigma}\ne0$, choose a nonzero quotient class in it. Since $\mathfrak H_{\rm sat}^{\Sigma}$ is the closed image of represented compactified saturation remainders, this class is represented by an attained remainder or by the declared compactified defect-measure class. It cannot have zero target capacity or zero certified target flux, because target-completeness would then place it in $\mathcal K_{\rm closed}^{\Sigma}$ and its quotient class would vanish. Thus it has positive target capacity and nonzero target flux. The legal-interface exhaustiveness hypothesis says every budget-admissible saturated channel decomposes into its closed-ledger part plus this target-visible quotient; hence an unlisted third channel would contradict the quotient construction. With the realization/nonpolarity certificate, positive target-visible-channel capacity is represented by a budget-admissible path reaching $\Sigma$. The two alternatives are mutually exclusive because the quotient class is either zero or nonzero. $\square$

### Theorem 11.23 (Constraint-Tail Calculation Equals Tail Cohomology)

### Definition 12.1 (Admissible Wiener Lift)
Let $H$ be a separable Hilbert state space and $\Omega$ a Wiener path space with presentation-fixed Wiener measure $\gamma$. The lifted state space is

\widetilde X=H\times\Omega,
\qquad \Lambda\mu=\mu\otimes\gamma,

### Theorem 12.2 (Noise as Geom on the Lift)
If the Malliavin form
$$
\mathcal E^{\rm noise}(F,F)=\int_{H\times\Omega}\|\mathcal B(u)^*D_\omega F\|_{U_0}^2\,d(\mu\otimes\gamma)
$$
is closable, then the stochastic forcing is a symmetric nonnegative Geom channel on the lifted space.

**Proof.** The Malliavin derivative $D_\omega$ and its divergence adjoint generate the Ornstein-Uhlenbeck Dirichlet form on Wiener space. Multiplication by the Hilbert-Schmidt coefficient $\mathcal B(u)$ preserves nonnegativity and symmetry of the quadratic form. Closability gives a symmetric form, hence a Geom channel. $\square$

### Theorem 12.3 (Quenched Dichotomy)
Assume the lifted continuum signature is admissible for $\gamma$-almost every noise realization. Then the deterministic channel and reaching dichotomy applies pathwise. In particular, if the lifted budget makes all reaching fixed Caus profiles unaffordable and residual flux is barred, then
$$
{\rm Cap}^{\omega}_{\mathcal B}(\Sigma)=0
\quad\text{for }\gamma\text{-a.e. }\omega.
$$

**Proof.** Freezing $\omega$ leaves a deterministic budget-admissible generator on the lifted space. The noise contribution remains in the symmetric form and only increases the charged Geom energy. The deterministic capacity and reaching theorems then apply for each budget-admissible path, outside a $\gamma$-null exceptional set. $\square$

### Definition 12.4 (Annealed Moment Capacities)
Let $\rho_t={\rm Law}(u_t)$ evolve under the forward generator on probability measures. For a represented family of moment observables $M_p$, define a moment-indexed form $\mathcal E^{(p)}$ and capacity
$$
{\rm Cap}^{(p)}_{\mathcal B}(\Sigma)
=\inf\{\mathcal E^{(p)}(F,F):F\ge1\text{ near the law-level target}\}.
$$

### Theorem 12.5 (Moment Spectrum)
Suppose the moment capacities are monotone in the sense that
$$
p_1\le p_2\quad\Rightarrow\quad {\rm Cap}^{(p_1)}_{\mathcal B}(\Sigma)\le {\rm Cap}^{(p_2)}_{\mathcal B}(\Sigma).
$$
Then the zero-capacity orders form a down-set and therefore have a critical order
$$
p_* = \sup\{p:{\rm Cap}^{(p)}_{\mathcal B}(\Sigma)=0\}.
$$
For $p<p_*$, zero capacity certifies that the $p$th moment target is unreached. For $p>p_*$, positive capacity only means that the zero-capacity exclusion has ended; actual reaching requires a separate nonpolarity or starting-law lower-bound certificate. This is the general intermittent-moment exclusion spectrum.

**Proof.** Monotonicity makes the zero-capacity set a down-set in the moment order: if ${\rm Cap}^{(p)}_{\mathcal B}=0$ and $p_1\le p$, then ${\rm Cap}^{(p_1)}_{\mathcal B}=0$. Its supremum is $p_*$. The capacity-polar theorem applies to each fixed $p$ on the law space, giving exclusion exactly at zero-capacity orders. Positive capacity is the negation of this exclusion certificate, not by itself a hitting lower bound. $\square$

### Theorem 12.6 (Lift Preserves Moment Observables)
For every base observable $h$ and every lifted law $\widetilde\mu$ with $\pi_*\widetilde\mu=\rho$,

\int_{H\times\Omega} h\circ\pi\,d\widetilde\mu=\int_H h\,d\rho.

Hence the lift does not create target or moment mass; it only changes the Geom conductance through the noise form.

**Proof.** This is the pushforward identity for $\pi_*\widetilde\mu=\rho$. Applying it to indicators gives target-fraction preservation; applying it to moment observables gives moment preservation. $\square$


## 12b. Conditional Expectations, Noncommutative Observable Algebras, and Expectation Gauges

This section imports `expectation_gauges.md` in a form that is independent of any particular stochastic PDE. The algebraic interface is allowed to be noncommutative. Abelian observable algebras appear as the commutative special case, and gauge reduction is represented by conditional expectation onto a fixed-point or readout algebra. The framework reads only represented conditional expectations, represented target readouts, and represented fluctuation-resistance or moment-capacity transcripts.

### Definition 12b.1 (Represented Observable Algebra and Conditional Expectation)
A noncommutative observable interface consists of:

1. a $\sigma$-finite von Neumann algebra $\mathcal M$ with faithful normal state or trace $\varphi$;
2. a budget-generated represented $*$-subalgebra $\mathcal M_{\mathcal B}\subset\mathcal M$ that is dense in the represented topology used by the problem;
3. an abelian target readout algebra $\mathcal C_\Sigma\subset\mathcal M$ generated by affordable target shells or target indicators;
4. a represented gauge, center, fixed-point, or coarse readout subalgebra $\mathcal N\subset\mathcal M$;
5. a normal unital completely positive idempotent
   $$
   E_{\mathcal N}:\mathcal M\to\mathcal N
   $$
   that is $\varphi$-preserving and $\mathcal N$-bimodular.

The map $E_{\mathcal N}$ is part of the problem certificate. In a finite or tracial algebra it is the canonical trace-preserving conditional expectation. In the general faithful normal state case its existence is guaranteed, for example, when $\mathcal N$ is invariant under the modular automorphism group of $\varphi$; otherwise it must be supplied as represented data. When $\mathcal M$ is abelian, this definition reduces to classical conditional expectation onto a sub-$\sigma$-algebra.

### Lemma 12b.2 (Conditional Expectation is the Gauge Projection)
Let $L^2(\mathcal M,\varphi)$ be the GNS Hilbert space. The conditional expectation $E_{\mathcal N}$ extends to a contractive orthogonal projection
$$
P_{\mathcal N}:L^2(\mathcal M,\varphi)\to L^2(\mathcal N,\varphi|_{\mathcal N}).
$$
For every $a\in L^2(\mathcal M,\varphi)$,
$$
a=P_{\mathcal N}a+(I-P_{\mathcal N})a,
$$
and the second term pairs to zero with every represented readout in $\mathcal N$.

**Proof.** A $\varphi$-preserving conditional expectation is contractive in the GNS $L^2$ norm by Kadison's inequality:
$$
\varphi(E_{\mathcal N}(a)^*E_{\mathcal N}(a))
\le \varphi(E_{\mathcal N}(a^*a))=\varphi(a^*a).
$$
It is idempotent with range $\mathcal N$. If $b\in\mathcal N$, bimodularity and $\varphi$-preservation give
$$
\varphi\bigl(b^*(a-E_{\mathcal N}a)\bigr)
=\varphi\bigl(E_{\mathcal N}(b^*a)-b^*E_{\mathcal N}a\bigr)=0.
$$
Thus the range is orthogonal to the kernel in $L^2$, and the extension is the orthogonal projection. $\square$

### Theorem 12b.3 (Noncommutative/Abelian Exhaustiveness by Expectation)
In a represented observable algebra interface, the decomposition
$$
a=E_{\mathcal N}a+(I-E_{\mathcal N})a
$$
has no third target-relevant component. The first term is represented structure in the gauge/readout algebra. The second term is residual for all affordable target readouts contained in $\mathcal N$. If it has a represented nonzero pairing with an affordable target shell outside $\mathcal N$, that pairing is a paid affordable witness and the corresponding shell/readout is added to the represented algebra before the verdict is read.

**Proof.** The first term lies in $\mathcal N$ by definition. Lemma 12b.2 shows that the second term pairs to zero with every represented readout in $\mathcal N$, hence has zero target flux for those target shells and belongs to the residual zero-target ledger. If a target shell outside $\mathcal N$ has nonzero represented pairing with $(I-E_{\mathcal N})a$, then that shell and scalar pairing are represented target data. By the affordable target-completeness rule, represented target data cannot remain residual: it is moved into the paid or public observable ledger with cost charged to $\mathcal B$. These alternatives exhaust the decomposition. $\square$

### Definition 12b.4 (Exact Gauge and Expectation Gauge)
Let $\widetilde L_\omega$ be a quenched generator on the lifted interface, let
$$
\bar L=\mathbb E_\omega[\widetilde L_\omega]
$$
be the annealed generator on a common represented core, and let $\bar{\mathcal E}^{\mathsf s}$ be the annealed symmetric form. A represented direction $\xi$ is an exact gauge when
$$
\widetilde L_\omega\xi=0
\quad\text{for }\gamma\text{-a.e. }\omega,
\qquad
\xi\in\ker\mathcal E_\omega^{\mathsf s}
\quad\text{for }\gamma\text{-a.e. }\omega,
$$
and the direction has zero affordable target readout. It is quotiented pathwise.

A represented direction $\xi$ is an expectation gauge when
$$
\bar L\xi=0,
\qquad
\delta_\omega\xi:=\widetilde L_\omega\xi-\bar L\xi
$$
is square-integrable, mean zero, and not identically zero on a set of positive $\gamma$-measure. It is annealed-exact but may have quenched leakage.

### Theorem 12b.5 (Annealed Closure of Expectation Gauges)
Assume $\xi$ is an expectation gauge, $\xi\in\ker\bar{\mathcal E}^{\mathsf s}$, and $\xi$ is invariant for the annealed gauge/readout algebra. Then $\xi$ is an exact gauge for the annealed generator and is placed in the annealed closed ledger. Moreover, for every continuous affordable target-flux functional $\ell_{\Sigma}$,
$$
\mathbb E_\omega\,\ell_{\Sigma}(\delta_\omega\xi)=0.
$$
Hence the fluctuation current is $J_{\rm res}^{0\Sigma}$ for the annealed verdict.

**Proof.** The equality $\bar L\xi=0$ is the defining mean-conservation condition. The kernel and invariance hypotheses make $\xi$ an exact gauge of the annealed form, so Definition 11.20 places it in the closed ledger as a gauge class with zero target readout. Since $\delta_\omega\xi$ is integrable and mean zero, linearity and continuity give
$$
\mathbb E_\omega\,\ell_{\Sigma}(\delta_\omega\xi)
=\ell_{\Sigma}(\mathbb E_\omega\delta_\omega\xi)=0.
$$
Target-completeness for the annealed algebra puts zero-target-flux residual currents in $J_{\rm res}^{0\Sigma}$. $\square$

### Corollary 12b.10 (Expectation Gauge Normal Form)
Expectation gauges add no primitive channel. At the annealed level they are exact gauges plus annealed-residual mean-zero currents. At the quenched level their only additional required data are the represented fluctuation-resistance or moment-capacity transcript. Any nonzero target-visible leak is either represented structure charged to the budget or the named fluctuation quotient; target-invisible leakage is $J_{\rm res}^{0\Sigma}$.

**Proof.** The annealed statement is Theorem 12b.5. The quenched statement is Theorem 12b.7 plus target-completeness. The algebraic no-fourth-channel statement is Theorem 12b.3 applied to the conditional expectation/gauge algebra. $\square$

## 13. Computable Certification Pipeline

A concrete application of the general continuum theorem requires represented hypotheses and finite-precision proof certificates. A concrete implementation produces the following transcript.

1. **Represent the state/interface.** Declare the state space approximation, public observables, reference measure, and target or saturated hull.
2. **Represent the generator form.** Give discrete or finite-element matrices/forms for $\mathcal E^{\mathsf s}$, the antisymmetric part, killing/boundary terms, and the sector-bound estimate.
3. **Build the quotient defect.** Construct $q$, $U$, $L_XU$, $UL_Q$, and the defect vector/current $E_q$.
4. **Solve the routing certificate.** Solve $\mathcal L_W\phi=E_q^{\rm rt}$ on the positive spectral subspace and report $\mathcal R_W(E_q)=\langle E_q,\mathcal L_W^+E_q\rangle$ plus the harmonic residual.
5. **Use the boundary-defect fast-track.** Estimate the structural gap $\gamma_{\rm str}$; if it is present, Theorem 4.3 gives closed range and therefore $\mathfrak H_{\rm bdry}^{\Sigma}=0$. If the gap is absent, provide the affordable target compactification, the compactified in-window boundary-defect space, the target-complete ledger $\mathcal K_{\rm bdry,closed}^{\Sigma}$, the quotient $\mathfrak H_{\rm bdry}^{\Sigma}$, and either a proof that it vanishes or a realization/nonpolarity certificate for the nonzero target-visible class. Escaping classes require no concentration-compactness proof: Lemma 13.6 places them in $J_{\rm res}^{0\Sigma}$. An explicit rigid/cost stratification is accepted only as a proof that the in-window quotient vanishes.
6. **Compute capacities.** Approximate equilibrium potentials for neighborhoods of $\Sigma$ and verify whether their affordable energies tend to zero, including any quotient-pullback correction term.
7. **For committor tasks.** Solve the killed or boundary-value problem, compute the residual $Lu$, and certify it by effective resistance.
8. **For SCRC tasks.** Provide $\varepsilon_*$, $\mathfrak C$, $\mathcal J$, P1-P3 proofs with uniform window constants, the in-window rigidity/cost certificate, and the rigid/cost exclusion transcript as ledger generators for $\mathfrak H_{\rm bdry}^{\Sigma}$. Escaping classes are closed by Lemma 13.6, which puts them in $J_{\rm res}^{0\Sigma}$.
9. **For conservation fast-track tasks.** Provide the carrier identity and window-independent remainders after constraint-tail closure.
10. **For elliptic constraint-tail tasks.** Provide the compactified homogeneous tail space $\mathcal T_{\rm tail}$, the target-complete tail ledger $\mathcal K_{\rm tail,closed}^{\Sigma}$, the quotient $\mathfrak H_{\rm tail}^{\Sigma}$, and a proof that it vanishes or a realization/nonpolarity certificate for the nonzero tail class. In elliptic finite-low-mode settings this transcript is the local inverse bound, homogeneous-tail proof on the core, dyadic summability constants, finite low-mode basis, gauge map matrix $\Gamma:\mathfrak g\to\mathcal H_{\rm low}$, and rank/cokernel computation.
11. **Propagate ledgers before later checks.** After each vanishing quotient, update the closed ledger by Definition 11.26 and declare the quotient-compatible transports into the next module. Boundary-defect generators are propagated before the tail quotient; boundary and tail generators are propagated before the flux radius; boundary, tail, and saturated generators are propagated before any final rigidity or capacity conclusion. The certificate must name the reduced space on which each later invariant is computed.
12. **Audit every remaining calculation by normal form.** Every explicit estimate, case split, rigidity argument, or equality classification must be registered as one of the three normal forms in Definition 11.29. The transcript must name its invariant: a target-complete quotient $\mathfrak H_{\mathcal F}^{\Sigma}$, a sharp radius $\alpha_{\mathcal F}$, or an equality-rigidity quotient $\mathfrak H_{\rm eq}^{\Sigma}$ with an optional separator. Theorem 11.30 is the admissibility check that the explicit calculation and the fast-track invariant are equivalent.
13. **For flux-sign tasks.** Provide the propagated reduced profile space $\mathcal P_{\rm crit}^{\rm red}$, the descended forms $\mathfrak C^{\rm red}$ and $\Phi_+^{\rm red}$, a certified value or upper bound for $\beta_*^{\rm red}$, and the descended operator $A^{-1/2}BA^{-1/2}$ in the quadratic case. When $\beta_*^{\rm red}=1$, also provide the certified compactification of normalized extremizing sequences in the reduced space, the saturation ledger $\Delta_{\rm flux}=\sum_aD_a$, the saturation remainder map $\mathscr R_{\rm sat}$, the target-complete closed ledger $\mathcal K_{\rm closed}^{\Sigma}$ with $J_{\rm res}^{0\Sigma}$ and every zero-capacity class included, the quotients $\mathcal O_{\rm ext}$, $\mathcal O_{\rm sat}$, and $\mathfrak H_{\rm sat}^{\Sigma}$, and either a proof that $\mathfrak H_{\rm sat}^{\Sigma}=0$ or a realization/nonpolarity certificate for the nonzero target-visible class. A Pohozaev separator may be used to prove $\mathfrak H_{\rm sat}^{\Sigma}=0$ by positivity on the target-visible quotient.
14. **Use the real conservative fast-track chain.** For conservative PDEs, assemble $\mathfrak H_{\rm bdry}^{\Sigma}=0$, propagate the boundary ledger, assemble $\mathfrak H_{\rm tail}^{\Sigma}=0$, propagate the tail ledger, and then check either $\beta_*^{\rm red}<1$ or $\beta_*^{\rm red}=1$ with $\mathfrak H_{\rm sat}^{\Sigma}=0$ on the propagated reduced space. Theorems 11.25, 11.27, and 11.30 prove this gives the same verdict as the explicit tail/sign/borderline calculation while forbidding repeated estimates of classes contained in the propagated ledger.
15. **For stochastic tasks.** Provide the Wiener lift, closability of the noise form, pathwise admissibility, quenched or annealed flux radius, martingale-control certificate, and any moment-capacity monotonicity certificate.
16. **For expectation-gauge / noncommutative algebra tasks.** Provide the represented von Neumann algebra or abelian algebra, faithful state, conditional expectation onto the gauge/readout algebra, exact or expectation-gauge direction, annealed generator, mean-zero fluctuation current, transverse restoring form, fluctuation resistance, and moment-capacity transcript. Ambient noncommutative correlations outside represented target readouts are residual unless a paid target witness is supplied.

The output certificate is not a simulation trace. It is a finite proof transcript: represented operators, solved linear systems or variational certificates, spectral-gap/closed-range evidence, propagated ledgers, reduced quotient spaces, fast-track normal-form invariants, capacity estimates, flux-radius bounds, conditional-expectation/gauge transcripts, fluctuation-resistance or moment-capacity certificates, and explicit budget charges.



### Definition 13.1 (Estimate Coverage Transcript)
An estimate coverage transcript is a finite table whose rows contain:

1. the represented estimate or calculation required by the problem interface;
2. the fast-track object on the current propagated quotient;
3. the theorem proving that the fast-track object gives the same non-reaching/target-reaching verdict as the explicit calculation;
4. the ledger update, when the row closes a family of classes.

For the continuum pipeline in this notebook the required rows are:

| Required estimate or check | Fast-track object | Equivalence theorem |
|---|---|---|
| legal state, generator, closability, sector bound | interface registration data and sector constant | Theorems 1.2 and 3.2-3.3 |
| budget composition and affordability transfer | B1-B4 transfer certificate | Theorem 2.3 |
| closed projection and directed exhaustiveness | structural gap $\gamma_{\rm str}$ or $\mathfrak H_{\rm bdry}^{\Sigma}$ | Theorems 4.3, 4.6, 4.7 |
| regularized projection solve | ridge certificate with declared $\lambda$-bias; exact closure delegated to $\gamma_{\rm str}$ or $\mathfrak H_{\rm bdry}^{\Sigma}$ | Corollary 4.4 |
| committor/operator residual | effective resistance of the represented residual | Theorems 5.3, 5.4, 6.2, 6.3 |
| defect drift estimate | $\mathcal R_W(E)$ plus harmonic residual | Theorem 6.4 |
| target exclusion or quotient pullback | ${\rm Cap}_{\mathcal B}(\Sigma)$ and quotient-pullback capacity bound | Theorems 7.2 and 7.3 |
| symmetrizing gauge | represented density bounds for the gauge | Theorem 7.4 |
| affordable target completeness | affordable windows, target shells, escaping/in-window compactification, and final propagated ledger | Lemma 13.6 and Theorem 13.7 |
| scale-fixed reaching cost | sharp normalized budget gap $\delta_{\Sigma}>0$ | Theorems 8.2, 8.3 and Theorem 11.30 |
| boundary-defect stratification | $\mathfrak H_{\rm bdry}^{\Sigma}=0$ with stratification rows as ledger generators | Theorems 9.2, 9.3 and 11.30 |
| SCRC threshold, P1-P3, in-window rigidity/cost, rigidity | target-complete boundary ledger and $\mathfrak H_{\rm bdry}^{\Sigma}$ after escaping classes are ledgered as $J_{\rm res}^{0\Sigma}$ | Theorems 10.2, 10.3, 13.6 and 11.30 |
| conservative carrier remainders | quotient-compatible carrier ledger propagated before the reduced flux check | Theorems 11.25, 11.27, 11.30 |
| elliptic constraint tails | $\mathfrak H_{\rm tail}^{\Sigma}$ with dyadic decay and gauge cokernel in the ledger | Theorem 11.23 |
| flux sign | reduced passivity radius $\beta_*^{\rm red}$ | Theorems 11.24 and 11.28 |
| borderline equality branch | $\mathfrak H_{\rm sat}^{\Sigma}$ on the reduced equality compactification | Theorems 11.21, 11.28, 11.30 |
| rigidity after saturation | $\mathfrak H_{\rm eq}^{\Sigma}$ or a Pohozaev/Noether separator positive on the target-visible quotient | Theorems 11.19 and 11.30 |
| stochastic lift and noise form | admissible Wiener lift, closable noise form, quenched/annealed radii and capacities | Theorems 12.2-12.6 |
| conditional expectation / expectation gauge | represented conditional expectation, gauge projection, fluctuation resistance, and moment-capacity transcript | Lemma 12b.2 and Theorems 12b.3, 12b.5, 12b.7, 12b.9 |

### Theorem 13.2 (Estimate-Coverage Completeness)
Assume a legal continuum signature and an estimate coverage transcript satisfying Definition 13.1. Then every represented estimate required by the notebook's global regularity or exclusion pipeline has a corresponding fast-track object in the table. The explicit estimate and the fast-track object give the same result: the same closed verdict, the same target-visible quotient, or the same feeding/zero-cost/extremizing alternative. When a row closes a family of classes, the closed classes are propagated by Definition 11.26 and Theorem 11.27 before any later row is computed.

**Proof.** The first two rows are registration rows. They supply the represented form, sector control, and budget algebra on which later modules are defined; they are not downstream case analyses. The directed-exhaustiveness rows are Theorems 4.3, 4.6, and 4.7: a Poincare gap, or the vanishing of $\mathfrak H_{\rm bdry}^{\Sigma}$, is equivalent to the explicit boundary-defect repair. The regularized row is Corollary 4.4; it is a biased computable solve and is not used as an exact closure unless the exact closed-range or boundary quotient row also closes.

The committor, residual, drift, capacity, and gauge rows are Theorems 5.3-7.4. Each replaces an explicit analytic estimate by a represented resistance, capacity, or gauge-admissibility certificate and gives the same bound or polarity conclusion. The scale-cost, boundary-stratification, SCRC, carrier, tail, flux, borderline, and rigidity rows are covered by Theorems 8.2-11.30. These rows are exactly the three normal forms of Definition 11.29: class-closure quotients, sign/gap radii, and equality-rigidity quotients or separators. The stochastic row follows by applying the same objects on the Wiener lift, pathwise or after expectation, as in Theorems 12.2-12.6. The conditional-expectation and expectation-gauge row is Section 12b: the conditional expectation is the gauge projection, annealed expectation gauges close as exact gauges plus annealed residuals, and quenched leakage is measured by the fluctuation-resistance or moment-capacity transcript.

The coverage table includes every module invoked by Section 13, including the affordable target-completeness row that removes escaping profiles from the list of target-reaching alternatives. Therefore any further represented estimate in the registered pipeline either instantiates one of these rows or supplies a new problem-interface hypothesis outside this notebook. For rows that close classes, Theorem 11.27 identifies later reduced computations with unreduced computations modulo the propagated ledger, so no later row re-estimates a closed class. Hence the fast-track route and the explicit calculation route have identical conclusions for every required estimate in the registered pipeline. $\square$


### Definition 13.3 (Full Fast-Track Instantiation Sheet)
A problem instantiates the full fast-track by filling the following sheet. Rows marked optional are supplied only when the problem asks for that module. The sheet is evaluated in order, every closed row is propagated before the next row is computed, and row 18 is terminal: if no in-window target-visible quotient remains, Theorem 13.7 gives the theorem verdict.

| Item | Data to provide | Fast-track to compute | Success condition | Complementary branch | Equivalence |
|---|---|---|---|---|---|
| 1. Legal signature | $X,\mu,\mathcal I,q,E_q,\mathcal L_W,\mathcal B,\Sigma$ | registration in the continuum interface | all objects represented and public | the problem is not yet in the interface | Theorem 1.2 |
| 2. Generator/form | quasi-regular closed form, Beurling-Deny-LeJan split, sector constant | sector transcript | form closed and sector bound finite | no generator-level theorem triggers | Theorems 3.2-3.3 |
| 3. Budget | resource monoid, cost map, ceiling | B1-B4 transcript | B1-B4 hold | affordability is not transportable | Theorem 2.3 |
| 4. Quotient defect | $q,U,L_XU,UL_Q,E_q$ | represented defect | defect is in the declared geometry | no routed quotient certificate | Theorem 1.2 |
| 5. Directed exhaustiveness | structural-gradient map $G$ | $\gamma_{\rm str}$ or $\mathfrak H_{\rm bdry}^{\Sigma}$ | $\gamma_{\rm str}>0$ or $\mathfrak H_{\rm bdry}^{\Sigma}=0$ | positive boundary-defect quotient | Theorems 4.3, 4.6, 4.7 |
| 6. Routing | defect $E$ and geometry $\mathcal L_W$ | $\mathcal R_W(E)$ and harmonic residual | finite resistance and harmonic part closed or zero | non-routable harmonic/target-visible residual | Theorems 6.2-6.4 |
| 7. Capacity | target neighborhoods and affordable potentials | ${\rm Cap}_{\mathcal B}(\Sigma)$, with pullback corrections | zero capacity for exclusion | positive capacity is routed by the capacity/nonpolarity fast-track to assert reaching | Theorems 7.2-7.3 |
| 8. Committor/operator module, optional | boundary sets or operator target, approximation class | residual resistance and active spectral gap | finite residual certificate and projection error accounted | approximation/residual obstruction | Theorems 5.2-5.4 |
| 9. Scale-reaching module, optional | scale map, fixed/moving split, scale-stability proof | sharp fixed-profile gap $\delta_{\Sigma}$ | $\delta_{\Sigma}>0$ and residual barred | zero-cost fixed profile or target-visible residual | Theorems 8.2-8.3, 11.30 |
| 10. SCRC/boundary repair, optional | $\varepsilon_*,\mathfrak C,\mathcal J$, P1-P3, in-window rigidity/cost certificate, rigidity rows | $\mathfrak H_{\rm bdry}^{\Sigma}$ with escaping classes first ledgered as $J_{\rm res}^{0\Sigma}$ | quotient vanishes | target-visible in-window boundary-defect class | Theorems 9.2-10.3, 13.6, 11.30 |
| 11. Conservative carrier, optional | carrier identity, uniform constants, remainder transports | carrier ledger on the propagated quotient | all remainders quotient-compatible and ledgered | target-visible carrier remainder | Theorems 11.25, 11.27, 11.30 |
| 12. Constraint tail, optional | elliptic split, homogeneous tail space, dyadic constants, low-mode basis, gauge map | $\mathfrak H_{\rm tail}^{\Sigma}$ | quotient vanishes | target-visible tail class | Theorem 11.23 |
| 13. Ledger propagation | transports $T_i$ and quotient-compatible descended forms | $\mathcal K_i^{\Sigma}$ and reduced spaces | every later invariant descends to the reduced space | cannot reuse earlier closure until quotient compatibility is supplied | Theorems 11.26-11.28 |
| 14. Flux sign, optional | reduced profile space, $\mathfrak C^{\rm red}$, $\Phi_+^{\rm red}$ | $\beta_*^{\rm red}$ | $\beta_*^{\rm red}<1$, or enter the equality row | $\beta_*^{\rm red}>1$ feeding profile | Theorems 11.24, 11.28 |
| 15. Borderline equality, optional | equality compactification, defects $D_a$, saturation map, closed ledger | $\mathfrak H_{\rm sat}^{\Sigma}$ | quotient vanishes | target-visible saturated class | Theorems 11.21, 11.28, 11.30 |
| 16. Rigidity/separator, optional | Pohozaev, Noether, modulation, Liouville, or monotonicity identities on equality set | $\mathfrak H_{\rm eq}^{\Sigma}$ or a positive separator | equality quotient vanishes | target-visible equality-rigidity class | Theorems 11.19, 11.30 |
| 17. Stochastic lift, optional | Wiener lift, closable noise form, martingale control, pathwise or annealed forms | lifted capacity and quenched/annealed radii | deterministic rows hold pathwise or in expectation with martingales controlled | stochastic target-visible quotient or unregistered martingale-control datum routed before evaluation | Theorems 12.2-12.6 |
| 17b. Expectation gauge / algebra row, optional | represented $\mathcal M,\varphi,\mathcal N,E_{\mathcal N}$, gauge direction $\xi$, fluctuation current, restoring form | conditional expectation projection, $\mathcal R_{\xi}^{\rm fl}$, and $p_*^{\rm fl}$ | annealed gauge closes and quenched leak is routed or zero-capacity | target-visible positive-capacity quotient handled by the capacity/nonpolarity fast-track | Corollary 12b.11 |
| 18. Affordable target completeness | affordable observable algebra, affordable windows, target shells, restriction maps, target-flux functionals, escaping/in-window split, and final propagated ledger | $\mathcal K_{\rm final}^{\Sigma}$ target-completeness | no in-window target-visible quotient remains, so exclusion/regularity follows | nonzero final quotient is a named in-window target-reaching alternative | Lemma 13.6, Theorem 13.7 |

### Theorem 13.4 (Full Fast-Track Instantiation Theorem)
If a problem supplies a completed instantiation sheet and every required non-optional row through row 18 satisfies its success condition, then row 18 gives the framework conclusion from the fast-track objects alone. The conclusion is the same conclusion obtained by the explicit calculation that proves the corresponding estimates and case classifications one by one. If any row takes its complementary target-reaching branch, the fast-track object names the same obstruction that the explicit calculation would have to isolate: a nonzero quotient class, a feeding profile, a zero-cost profile, a non-routable harmonic residual, a positive-capacity target-visible class, or a stochastic target-visible quotient or unregistered martingale-control datum routed before evaluation.

**Proof.** Rows 1-4 register the problem and make the represented objects available. They are interface triggers. Rows 5-17b are exactly the estimate coverage rows of Definition 13.1, refined with the data needed to instantiate each invariant. By Theorem 13.2, each such row has the same result as the corresponding explicit estimate. When a row closes classes, Theorem 11.27 propagates those classes into the ledger before later rows are evaluated; the quotient isomorphism in Theorem 11.27 proves that later reduced computations are identical to unreduced computations modulo the closed classes. Row 18 is Theorem 13.7: escaping classes are residual by Lemma 13.6, every in-window zero-flux or zero-capacity class is in the propagated ledger, and the absence of a nonzero final quotient is exactly the theorem verdict. Therefore a successful sheet composes the same implications as the explicit calculation but replaces every casewise calculation by its quotient, radius, gap, resistance, capacity, or equality-rigidity invariant. If a row takes its target-reaching branch, the dichotomy theorem for that row gives the named target-reaching alternative, and target-completeness prevents an additional unlisted obstruction inside the registered interface. $\square$


### Definition 13.5 (Affordable Target Compactification and Flux)
Fix a legal continuum signature and budget $\mathcal B$. Let
$$
\mathcal F_{\rm obs}^{\mathcal B}
=\sigma\{\gamma:{\rm cost}(\gamma)\preceq b(n)\}
$$
be the affordable observable algebra. An affordable target compactification consists of:

1. a directed family of affordable windows $W_R\in\mathcal F_{\rm obs}^{\mathcal B}$ such that every affordable observable and every affordable structural test current is approximable in form norm by a current supported in some $W_R$;
2. affordable target shells $S_r$ around $\Sigma$ with $S_r\subset W_{R(r)}$ for some window and $S_r\in\mathcal F_{\rm obs}^{\mathcal B}$;
3. a compactified represented edge/channel space $\widehat{\mathcal H}_{\rm edge}^{\mathcal B}$ with continuous restriction maps
   $$
   \rho_R:\widehat{\mathcal H}_{\rm edge}^{\mathcal B}\to\mathcal H_{\rm edge}(W_R);
   $$
4. continuous target-flux functionals factoring through affordable shells,
   $$
   {\rm Flux}_{\Sigma,r}(F)=\ell_{\Sigma,r}(\rho_{R(r)}F);
   $$
5. the affordable structural closure
   $$
   \mathcal C^{\mathcal B}=\overline{{\rm im}\,G}^{\mathcal B}\subseteq\widehat{\mathcal H}_{\rm edge}^{\mathcal B}
   $$
   and the residual zero-target ledger generated by classes orthogonal to affordable structural tests and carrying zero affordable target flux.

A compactified class $F\in\widehat{\mathcal H}_{\rm edge}^{\mathcal B}$ is **escaping** when $\rho_R F=0$ for every affordable window $W_R$. It is **in-window** when $\rho_R F\ne0$ for at least one affordable window. A directed channel contributes budgeted target progress only through the represented fluxes ${\rm Flux}_{\Sigma,r}$, represented target readouts, or a represented positive-capacity/nonpolarity witness. A component with no such witness is target-invisible in the affordable algebra.

### Lemma 13.6 (Affordable-Flux Zero and Escaping-Residual Lemma)
Every escaping compactified class has zero flux through every affordable target shell and pairs trivially with every affordable structural-gradient test current. Hence it lies in the residual zero-target ledger $J_{\rm res}^{0\Sigma}$ and cannot reach $\Sigma$ through the affordable algebra.

More generally, if a represented or compactified class has zero affordable target flux and zero represented target readout after the declared projections, it is a residual zero-target class. If a nonzero represented pairing with an affordable target shell is supplied, that pairing is an affordable target witness; the class is then moved to the public or paid target ledger before the verdict is read.

**Proof.** Let $F$ be escaping. For every target shell $S_r$, choose $R(r)$ with $S_r\subset W_{R(r)}$. Since $\rho_{R(r)}F=0$ and the flux functional factors through this restriction,
$$
{\rm Flux}_{\Sigma,r}(F)=\ell_{\Sigma,r}(\rho_{R(r)}F)=0.
$$
Now let $G\Phi$ be an affordable structural-gradient test current. By affordable localization in Definition 13.5, for every $\varepsilon>0$ there is a current $H_R$ supported in some $W_R$ with $\|G\Phi-H_R\|_{\rm form}<\varepsilon$. The pairing with the localized current is zero because $\rho_R F=0$. Continuity of the compactified pairing gives
$$
|\langle F,G\Phi\rangle|
\le |\langle F,H_R\rangle|+|\langle F,G\Phi-H_R\rangle|
\le C_F\varepsilon.
$$
Letting $\varepsilon\downarrow0$ gives $\langle F,G\Phi\rangle=0$ for every affordable structural test current, and therefore for the affordable structural closure. Thus $F$ is residual and has zero target flux. The zero-flux residual bar and Theorem 7.2 exclude affordable reaching of $\Sigma$.

For the general statement, target progress is defined only through represented affordable target functionals. If all such functionals vanish on a class, the class is exactly a zero-target residual class in the declared interface. If one of them is nonzero, the corresponding scalar transcript is represented target data; by the budget rules it is charged to $\mathcal B$ and the class is read in the public or paid target ledger rather than in the residual ledger. These alternatives exhaust the affordable target algebra. $\square$

### Theorem 13.7 (Unconditional Affordable Target-Completeness Theorem)
Assume a legal continuum signature, an affordable target compactification, and a successful fast-track transcript for all required rows before the final affordable-completeness audit. Let $\mathcal K_{\rm final}^{\Sigma}$ be the propagated closed ledger after those successful rows have been applied. Then $\mathcal K_{\rm final}^{\Sigma}$ is target-complete in the affordable algebra:

1. every escaping class lies in $J_{\rm res}^{0\Sigma}$ by Lemma 13.6;
2. every represented or compactified in-window class with zero affordable target flux, or zero budget target capacity, lies in $\mathcal K_{\rm final}^{\Sigma}$;
3. every class in $\mathcal K_{\rm final}^{\Sigma}$ is certified non-reaching by a ledger generator, zero-capacity certificate, zero-flux residual certificate, quotient-vanishing theorem, sign/gap certificate, or equality-rigidity separator;
4. every nonzero class in the final quotient has positive affordable target capacity and nonzero affordable target flux, and is one of the named target-reaching alternatives in Definition 13.3.

Consequently the fast-track has no additional concentration-compactness or escaping-profile obligation. The only target-visible classes left after Lemma 13.6 are in-window classes, and those are exactly the classes measured by the registered quotient, radius, gap, resistance, capacity, conditional-expectation/fluctuation, passivity, saturation, and equality-rigidity rows. For boundary-defect rows with SCRC data, Theorem 10.2 gives the in-window rigid/cost dichotomy; escaping classes are not part of that dichotomy because Lemma 13.6 places them in $J_{\rm res}^{0\Sigma}$.

**Proof.** Let $F$ be any represented or compactified class in the final channel space. By Definition 13.5, either $F$ is escaping or it is in-window.

If $F$ is escaping, Lemma 13.6 places it in $J_{\rm res}^{0\Sigma}$. This residual zero-target class is a generator of every target-complete closed ledger used in Sections 4, 11, and 13, and it is propagated by Theorem 11.27. Hence $F\in\mathcal K_{\rm final}^{\Sigma}$. Escaping profiles therefore cannot form a fourth target-visible category.

Assume now that $F$ is in-window. Then some restriction $\rho_R F$ is visible to the affordable observable algebra and to the relevant represented module. If $F$ has zero budget target capacity, the zero-capacity row of Definitions 4.5, 11.20, and 13.1 puts it in the closed ledger; lower semicontinuity preserves this inclusion under closure. If $F$ has zero target flux and no represented target readout, Lemma 13.6 places it in the residual zero-target ledger. If a represented readout is nonzero, that readout is a paid or public target witness and the class is no longer residual; it is evaluated by the row whose invariant detects it.

It remains to identify the possible in-window target-visible rows. The estimate-coverage transcript of Definition 13.1 and the instantiation sheet of Definition 13.3 enumerate them: structural-gradient routing, boundary-defect quotient, capacity, committor/resistance, scale-reaching, SCRC, conservative carrier, constraint tail, flux radius, borderline saturation, equality-rigidity, stochastic lift, and expectation-gauge/algebra rows. For a class-closure row, a successful quotient-vanishing certificate places all zero-flux or zero-capacity classes in the row ledger. For a radius/gap row, a successful bound is the represented sign or gap certificate for the descended in-window quotient. For an equality-rigidity row, the positive separator or vanishing equality quotient places the class in the ledger. Theorem 11.27 propagates every successful row ledger into $\mathcal K_{\rm final}^{\Sigma}$ before later invariants are computed.

The converse non-reaching statement for ledger elements follows from the construction of the ledger. Its generators are exactly represented structural gradients routed or charged by their row certificate, residual zero-flux classes, gauge or symmetry directions with zero target readout, dyadic/rebased/subthreshold classes, rigidly excluded classes, Pohozaev/Noether/separator-killed classes, sign/gap-closed classes, and zero-capacity classes. Each generator is non-reaching by the theorem cited in the corresponding row of Definitions 13.1 and 13.3; closedness and lower semicontinuity preserve non-reaching under represented limits.

Finally, suppose a class survives in the quotient by $\mathcal K_{\rm final}^{\Sigma}$. It is not escaping by the first paragraph. It cannot have zero target flux or zero budget target capacity by the previous paragraphs. Therefore it has positive affordable target capacity and nonzero affordable target flux. The estimate-coverage and instantiation theorems, Theorems 13.2 and 13.4, identify the row in which it survives: it is a positive boundary/tail/saturation/equality quotient, a feeding profile, a zero-cost profile, a non-routable harmonic residual with target-visible witness, a positive-capacity target-visible class, or a stochastic/fluctuation target-visible quotient or unregistered martingale-control datum routed before evaluation. These are exactly the named target-reaching alternatives. Thus the propagated fast-track ledger is exhaustive and target-complete, and the fast-track gives the same result as the explicit calculation without any separate concentration-compactness step. $\square$

### Corollary 13.8 (Measured Dynamical Target Specialization)

A finite or exhausted measured dynamical target problem is a specialization of the continuum target-completeness spine when it supplies:

1. a represented transition kernel, generator, or deterministic map;
2. an initial law and reference law;
3. represented target shells and a time, height, or denominator budget;
4. a committor/capacity transcript for the target shells;
5. a law-space fluctuation, martingale bracket, or covariance transcript when the problem is stochastic or quenched;
6. a tail-event transcript for finitely-many or infinitely-many target hits;
7. a Foster-Lyapunov drift transcript when recurrence, escape, or absorption is certified by drift;
8. a realization/nonpolarity policy for positive target witnesses.

Under these hypotheses the measured dynamical target row has no independent analytic estimate outside the existing fast-track objects. Its invariant is the finite tuple consisting of target capacity, recurrence quotient, law-space fluctuation radius, drift defect, tail-event ledger, and realization map.

**Proof.** A finite kernel is a Markov generator on the represented finite Hilbert space, and an exhausted kernel is handled window by window with the declared limit transcript. The symmetric part supplies the capacity form, while the directed part is a represented Caus channel. Committors are the hitting-law quotients of Section 5, target polarity and positive capacity are the rows of Section 7, stochastic and expectation components are the rows of Sections 12 and 12b, and escaping path classes are placed in the residual zero-target ledger by Lemma 13.6. The Foster-Lyapunov transcript is a represented drift inequality and hence a row invariant in the sign/gap normal form of Definition 11.29. Tail events are target-shell statements in the path lift; summable tails enter the zero-target ledger, and divergent quasi-independent tails are measured by the law-space covariance transcript. Theorem 13.7 then gives target-completeness for the propagated ledger. Thus the measured dynamical target specialization is the ordered composition of the registered rows named in the preceding sentences. $\square$


## 14. Summary

The general continuum extension has one spine. A legal continuum signature registers a quasi-regular form, public observables, a represented quotient defect, a defect geometry, a budget, and a target. The budget axioms make affordability transferable. The continuum generator decomposes into symmetric, antisymmetric, and boundary parts under the sector condition. The directed Hodge part is exhaustive exactly when the structural-gradient range is closed, equivalently when the structural Laplacian has a spectral gap. If that range is not closed, the boundary-defect fast-track computes the target-complete cohomology $\mathfrak H_{\rm bdry}^{\Sigma}$; the rigid/cost stratification is one proof that this quotient vanishes. Theorem 4.7 proves the quotient fast-track and the explicit non-fast-track repair give the same non-reaching/target-reaching verdict.

Effective resistance is the routing certificate in the coercive regime and the Poincare face of the full defect-geometry inequality profile: $\mathcal R_W(E)=\langle E,\mathcal L_W^+E\rangle$ where finite, with harmonic residual as the non-routable obstruction. When Geom is degenerate, commutator routing replaces the direct inverse by the positive bracket closure in hypocoercive systems or by a Mourre/Morawetz positive-commutator carrier in conservative dispersive systems. Capacity is the profile face that turns routing into reachability: zero budget capacity makes the target polar and unreachable. A finite dissipation budget excludes scale-fixed Caus profiles when the scale-stability certificate is present and each fixed reaching profile carries positive per-scale cost over infinitely many log-scale steps. SCRC packages gapless closure as ledger generators for $\mathfrak H_{\rm bdry}^{\Sigma}$, and the conservation fast-track reduces P1-P3 to a carrier identity with uniform constants.

For conservative PDEs the real fast-track chain is
$$
\mathfrak H_{\rm bdry}^{\Sigma}=0,
\qquad
\mathfrak H_{\rm tail}^{\Sigma}=0,
\qquad
\beta_*^{\rm red}<1
\quad\text{or}\quad
\bigl(\beta_*^{\rm red}=1\text{ and }\mathfrak H_{\rm sat}^{\Sigma}=0\bigr).
$$
The vanished quotients are propagated into the closed ledger before the next invariant is computed. Theorem 11.27 is the formal rule: quotient-compatible later checks on the reduced space are canonically identical to unreduced checks modulo all previously closed classes, by the third isomorphism theorem for closed ledgers. Corollary 11.28 applies this to the conservative chain, so tail, sign, borderline, Pohozaev, rigidity, and capacity checks never re-estimate boundary, tail, residual, gauge, or saturated classes contained in earlier propagated ledgers.

Theorem 11.25 proves that this chain gives the same non-reaching/target-reaching verdict as the explicit non-fast-track calculation that estimates every constraint tail, proves every flux sign inequality, and classifies every borderline equality remainder. Theorem 11.30 gives the full normal form for any remaining represented calculation: it must be a target-complete quotient, a sharp radius or gap, or an equality-rigidity quotient with an optional separator. Corollary 11.31 lists the registered fast-track invariant for every calculation class used by the pipeline. Definitions 13.1, 13.3, and 13.5 and Theorems 13.2, 13.4, and 13.7 make this row-by-row: every required estimate has a named fast-track object, explicit instantiation data, a success condition, a target-reaching alternative, and an equivalence theorem proving that the fast-track and the explicit calculation give the same result. The affordable target-completeness theorem is the global closure rule. It removes escaping profiles from the list of target-reaching alternatives by placing them in $J_{\rm res}^{0\Sigma}$ through the affordable-flux zero lemma, and it leaves only in-window target-visible classes to be measured by the registered fast-track invariants. A target-complete closed ledger puts $J_{\rm res}^{0\Sigma}$, gauge, exact/routed, dyadic, rebased, subthreshold, Pohozaev-killed, rigidly excluded, and zero-capacity channels into the closed side. Nonzero target-complete quotients or $\beta_*>1$ are the corresponding positive feeding/reaching alternatives, subject to realization/nonpolarity. Stochastic problems enter by an admissible Wiener lift, where noise is Geom, plus quenched or annealed versions of the same flux-radius and capacity certificates.


Noncommutative observable algebras and abelian readout algebras enter through represented conditional expectations. A $\varphi$-preserving conditional expectation is the $L^2$ gauge projection; its range is charged structure and its orthogonal complement is residual for every represented readout in the gauge algebra. Expectation gauges are exact gauges for the annealed generator, and their mean-zero fluctuation currents are annealed $J_{\rm res}^{0\Sigma}$. The quenched obligation is not a new channel: it is the represented fluctuation-resistance or moment-capacity row. Finite fluctuation resistance routes the leak; zero capacity or zero target flux closes it; positive capacity is routed by the capacity/nonpolarity fast-track. The fluctuation sign is irreducible only relative to deterministic annealed data because it depends on covariance data erased by annealing.

No theorem uses an argument outside the registered interface. A PDE or continuous domain is registered by the signature, and each metatheorem triggers only after the corresponding certificate transcript in Section 13 is supplied. Within this registered interface, there is no independent calculation class outside the fast-track normal forms of Definition 11.29, the estimate coverage transcript of Definition 13.1, the instantiation sheet of Definition 13.3, and the affordable target-completeness theorem.
